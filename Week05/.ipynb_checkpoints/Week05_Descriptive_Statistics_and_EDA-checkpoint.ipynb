## 1. Types of Data

Before computing statistics, we must understand **what kind of data** we have.  
The type determines which statistics and visualizations are appropriate.

| Category | Type | Description | Examples | Statistics |
|----------|------|-------------|----------|------------|
| **Numerical** | Continuous | Can take any value in a range | Height, GPA, Temperature | Mean, Std, Correlation |
| **Numerical** | Discrete | Countable integers | Number of courses, Age in years | Mean, Median, Mode |
| **Categorical** | Nominal | Categories with no order | Major, Gender, Color | Mode, Frequency |
| **Categorical** | Ordinal | Categories with meaningful order | Grade (A>B>C), Satisfaction | Median, Mode, Frequency |

### Why Does It Matter?

- Computing the **mean** of zip codes is meaningless (nominal data)
- A **bar chart** is correct for majors, but a **histogram** is correct for scores
- **Correlation** only applies to numerical data

Let's create a sample dataset to use throughout this notebook. We'll simulate a class of 80 students.

````python
# Generate a realistic student dataset
rng = np.random.default_rng(2026)
n = 80

df = pd.DataFrame({
    "Name":      [f"Student_{i+1:03d}" for i in range(n)],
    "Age":       rng.integers(18, 26, size=n),
    "Gender":    rng.choice(["Male", "Female"], size=n),
    "Major":     rng.choice(["Data Science", "Statistics", "Mathematics", "CS", "Economics"], size=n),
    "Year":      rng.choice(["Freshman", "Sophomore", "Junior", "Senior"], size=n, p=[0.15, 0.3, 0.35, 0.2]),
    "Midterm":   rng.normal(72, 14, size=n).clip(0, 100).round(1),
    "Final":     rng.normal(70, 16, size=n).clip(0, 100).round(1),
    "Homework":  rng.normal(82, 10, size=n).clip(0, 100).round(1),
    "Attendance": rng.integers(50, 101, size=n),  # percentage
    "StudyHours": rng.exponential(8, size=n).clip(1, 40).round(1),  # hours per week
})

# Add some missing values (realistic)
df.loc[rng.choice(n, 4, replace=False), "Midterm"] = np.nan
df.loc[rng.choice(n, 3, replace=False), "Final"] = np.nan
df.loc[rng.choice(n, 2, replace=False), "Attendance"] = np.nan

# Save for reuse
df.to_csv("students_eda.csv", index=False)

print(f"Dataset: {df.shape[0]} students × {df.shape[1]} columns")
df.head(10)
````

Check the data types — Pandas automatically infers types, but we should verify they match our expectations.

````python
# Separate numerical and categorical columns
print("=== Data Types ===")
print(df.dtypes)

# Identify column types
num_cols = df.select_dtypes(include="number").columns.tolist()
cat_cols = df.select_dtypes(include="object").columns.tolist()

print(f"\nNumerical columns ({len(num_cols)}): {num_cols}")
print(f"Categorical columns ({len(cat_cols)}): {cat_cols}")
````

## 2. Measures of Central Tendency

Central tendency answers: **"What is a typical value?"**

| Measure | Formula | Best For | Sensitive to Outliers? |
|---------|---------|----------|------------------------|
| **Mean** | $\bar{x} = \frac{1}{n}\sum x_i$ | Symmetric data | Yes — very sensitive |
| **Median** | Middle value when sorted | Skewed data | No — robust |
| **Mode** | Most frequent value | Categorical data | No |

### When to Use Which?

- **Income data** → Median (skewed by billionaires)
- **Exam scores** → Mean (roughly symmetric)
- **Favorite color** → Mode (categorical)

### 2.1 Mean

The arithmetic average — sum divided by count. Most commonly used, but pulled toward outliers.

````python
# Mean of each numeric column
print("=== Mean ===")
print(df[["Midterm", "Final", "Homework", "StudyHours"]].mean().round(2))

# Manual calculation to verify
midterm_mean = df["Midterm"].sum() / df["Midterm"].count()  # .count() skips NaN
print(f"\nManual Midterm mean: {midterm_mean:.2f}")
print(f"Pandas Midterm mean: {df['Midterm'].mean():.2f}")
````

### 2.2 Median

The middle value when data is sorted. Exactly 50% of values are above and 50% below.  
Much more robust to outliers than the mean.

````python
# Median of each numeric column
print("=== Median ===")
print(df[["Midterm", "Final", "Homework", "StudyHours"]].median().round(2))

# Compare mean vs median — large difference suggests skewness
print("\n=== Mean vs Median ===")
for col in ["Midterm", "Final", "StudyHours"]:
    mean_val = df[col].mean()
    med_val = df[col].median()
    diff = mean_val - med_val
    print(f"{col:12s}  Mean={mean_val:.1f}  Median={med_val:.1f}  Diff={diff:+.1f}")
````

### 2.3 Mode

The most frequently occurring value. Essential for categorical data; can have multiple modes.

````python
# Mode for categorical columns
print("=== Mode ===")
print(f"Most common Major:  {df['Major'].mode()[0]}")
print(f"Most common Gender: {df['Gender'].mode()[0]}")
print(f"Most common Year:   {df['Year'].mode()[0]}")

# Mode for numerical columns (less useful, but possible)
print(f"\nMost common Age:    {df['Age'].mode()[0]}")

# Frequency of the mode
print("\n=== Major Frequencies ===")
print(df["Major"].value_counts())
````

### 2.4 Demonstrating Outlier Sensitivity

Let's see how a single extreme value affects the mean vs the median.

````python
# Original data
salaries = pd.Series([35000, 42000, 45000, 48000, 50000])
print(f"Original:   Mean={salaries.mean():,.0f}  Median={salaries.median():,.0f}")

# Add one CEO salary
salaries_with_ceo = pd.concat([salaries, pd.Series([500000])], ignore_index=True)
print(f"With CEO:   Mean={salaries_with_ceo.mean():,.0f}  Median={salaries_with_ceo.median():,.0f}")

# The mean jumped from 44,000 to 120,000 — but the median barely moved!
print(f"\nMean increased by: {salaries_with_ceo.mean() - salaries.mean():,.0f}")
print(f"Median increased by: {salaries_with_ceo.median() - salaries.median():,.0f}")
````

## 3. Measures of Dispersion (Spread)

Central tendency alone is not enough — we also need to know **how spread out** the data is.

| Measure | Description | Formula |
|---------|-------------|--------|
| **Range** | Max − Min | $R = x_{max} - x_{min}$ |
| **Variance** | Average squared deviation from mean | $s^2 = \frac{1}{n-1}\sum(x_i - \bar{x})^2$ |
| **Standard Deviation** | Square root of variance (same units as data) | $s = \sqrt{s^2}$ |
| **IQR** | Range of the middle 50% | $IQR = Q_3 - Q_1$ |
| **Coefficient of Variation** | Relative spread (unitless) | $CV = \frac{s}{\bar{x}} \times 100\%$ |

> Note: Pandas uses $n-1$ (Bessel's correction) by default — this is the **sample** standard deviation.

### 3.1 Range

The simplest measure of spread — but very sensitive to outliers (uses only 2 data points).

````python
# Range for exam scores
for col in ["Midterm", "Final", "Homework"]:
    col_min = df[col].min()
    col_max = df[col].max()
    col_range = col_max - col_min
    print(f"{col:10s}  Min={col_min:.1f}  Max={col_max:.1f}  Range={col_range:.1f}")
````

### 3.2 Variance and Standard Deviation

Variance measures how far values are from the mean on average (in squared units).  
Standard deviation is the square root of variance — back in the **original units**, making it more interpretable.

````python
# Variance and standard deviation
for col in ["Midterm", "Final", "Homework", "StudyHours"]:
    var = df[col].var()   # sample variance (ddof=1 by default)
    std = df[col].std()   # sample std
    print(f"{col:12s}  Variance={var:8.2f}  Std Dev={std:.2f}")

# Manual calculation to understand the formula
midterm = df["Midterm"].dropna()
mean = midterm.mean()
manual_var = ((midterm - mean) ** 2).sum() / (len(midterm) - 1)
manual_std = np.sqrt(manual_var)
print(f"\nManual Midterm Std: {manual_std:.2f}")
print(f"Pandas Midterm Std: {midterm.std():.2f}")
````

### 3.3 Interquartile Range (IQR)

The IQR measures the spread of the **middle 50%** of data. It's robust to outliers because it ignores the extremes.

- **Q1** (25th percentile): 25% of data is below this value
- **Q2** (50th percentile): the median
- **Q3** (75th percentile): 75% of data is below this value
- **IQR** = Q3 − Q1

````python
# Quartiles and IQR for Midterm
midterm = df["Midterm"].dropna()

q1 = midterm.quantile(0.25)
q2 = midterm.quantile(0.50)  # = median
q3 = midterm.quantile(0.75)
iqr = q3 - q1

print("=== Midterm Quartiles ===")
print(f"Q1 (25%):  {q1:.1f}")
print(f"Q2 (50%):  {q2:.1f}  (= Median)")
print(f"Q3 (75%):  {q3:.1f}")
print(f"IQR:       {iqr:.1f}")

# Quick quartiles for all numeric columns using describe()
print("\n=== Percentiles for All Exam Scores ===")
print(df[["Midterm", "Final", "Homework"]].describe().loc[["25%", "50%", "75%"]])
````

### 3.4 Coefficient of Variation (CV)

CV expresses the standard deviation as a **percentage of the mean**. Useful for comparing variability across different scales.

````python
# Coefficient of Variation
print("=== Coefficient of Variation ===")
for col in ["Midterm", "Final", "Homework", "StudyHours", "Attendance"]:
    mean = df[col].mean()
    std = df[col].std()
    cv = (std / mean) * 100
    print(f"{col:12s}  CV = {cv:.1f}%")

# Higher CV → more relative variability
# StudyHours likely has high CV (some study 2h, some 30h)
````

## 4. Skewness & Kurtosis

Beyond center and spread, the **shape** of the distribution matters.

### Skewness — Asymmetry

| Skewness | Shape | Mean vs Median |
|----------|-------|----------------|
| **= 0** | Symmetric (normal) | Mean ≈ Median |
| **> 0** | Right-skewed (long right tail) | Mean > Median |
| **< 0** | Left-skewed (long left tail) | Mean < Median |

### Kurtosis — Tail Heaviness

| Kurtosis (excess) | Shape | Meaning |
|-------------------|-------|--------|
| **= 0** | Mesokurtic | Same tails as normal |
| **> 0** | Leptokurtic | Heavy tails, sharp peak |
| **< 0** | Platykurtic | Light tails, flat peak |

Compute skewness and kurtosis for each numeric column. Pandas uses **Fisher's definition** (excess kurtosis, where normal = 0).

````python
# Skewness and Kurtosis for all numeric columns
num_cols_exam = ["Midterm", "Final", "Homework", "StudyHours", "Attendance"]

shape_stats = pd.DataFrame({
    "Mean":     df[num_cols_exam].mean(),
    "Median":   df[num_cols_exam].median(),
    "Skewness": df[num_cols_exam].skew(),
    "Kurtosis": df[num_cols_exam].kurtosis(),
}).round(2)

print(shape_stats)

# Interpretation
print("\n=== Interpretation ===")
for col in num_cols_exam:
    sk = df[col].skew()
    if abs(sk) < 0.5:
        label = "approximately symmetric"
    elif sk > 0:
        label = "right-skewed (tail to the right)"
    else:
        label = "left-skewed (tail to the left)"
    print(f"  {col}: skew={sk:.2f} → {label}")
````

Visualize symmetric vs skewed distributions side by side.

````python
# Compare distributions with different skewness
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Left-skewed (artificial)
left_skew = 100 - rng.exponential(15, 500)
axes[0].hist(left_skew, bins=30, color="steelblue", edgecolor="white")
axes[0].set_title(f"Left-Skewed\n(skew={pd.Series(left_skew).skew():.2f})")
axes[0].axvline(np.mean(left_skew), color="red", label="Mean", linewidth=2)
axes[0].axvline(np.median(left_skew), color="orange", label="Median", linewidth=2)
axes[0].legend()

# Symmetric
symmetric = rng.normal(70, 15, 500)
axes[1].hist(symmetric, bins=30, color="mediumseagreen", edgecolor="white")
axes[1].set_title(f"Symmetric\n(skew={pd.Series(symmetric).skew():.2f})")
axes[1].axvline(np.mean(symmetric), color="red", label="Mean", linewidth=2)
axes[1].axvline(np.median(symmetric), color="orange", label="Median", linewidth=2)
axes[1].legend()

# Right-skewed
right_skew = rng.exponential(10, 500)
axes[2].hist(right_skew, bins=30, color="coral", edgecolor="white")
axes[2].set_title(f"Right-Skewed\n(skew={pd.Series(right_skew).skew():.2f})")
axes[2].axvline(np.mean(right_skew), color="red", label="Mean", linewidth=2)
axes[2].axvline(np.median(right_skew), color="orange", label="Median", linewidth=2)
axes[2].legend()

plt.tight_layout()
plt.show()
````

## 6. Frequency Tables & Cross-Tabulation

For categorical data, **frequency tables** are the equivalent of `describe()` — they show how often each category appears.

### 6.1 Value Counts

Count occurrences of each category — both as raw counts and as proportions (percentages).

````python
# Frequency table: counts
print("=== Major Distribution ===")
print(df["Major"].value_counts())

# Frequency table: proportions (percentages)
print("\n=== Major Distribution (%) ===")
print((df["Major"].value_counts(normalize=True) * 100).round(1))
````

### 6.2 Cross-Tabulation

`pd.crosstab()` shows the joint distribution of **two categorical variables** — like a pivot table of counts.

````python
# Cross-tabulation: Major × Gender
cross = pd.crosstab(df["Major"], df["Gender"], margins=True, margins_name="Total")
print("=== Major × Gender ===")
print(cross)

# With row percentages (each row sums to 100%)
print("\n=== Major × Gender (Row %) ===")
cross_pct = pd.crosstab(df["Major"], df["Gender"], normalize="index") * 100
print(cross_pct.round(1))
````

Cross-tab with **Year** — useful for understanding the student composition.

````python
# Cross-tabulation: Major × Year
cross_year = pd.crosstab(df["Major"], df["Year"], margins=True)
print("=== Major × Year ===")
print(cross_year)
````

## 8. Box Plots & the Five-Number Summary

A box plot visualizes the **five-number summary**: Min, Q1, Median, Q3, Max.  
It's the best single-chart tool for spotting **outliers, spread, and skewness**.

```
    ◄── 1.5×IQR ──►    ◄── IQR ──►    ◄── 1.5×IQR ──►
    |                   |         |                   |
    ├─────── whisker ───┤   box   ├─── whisker ───────┤
                        Q1   Med  Q3
                             │
    ○ ○                                          ○  ← outliers
```

### 8.1 Basic Box Plot

Box plots for all exam scores — compare spread and symmetry across exams.

````python
# Box plots for exam scores
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# All exams side by side
df[["Midterm", "Final", "Homework"]].boxplot(ax=axes[0])
axes[0].set_title("Score Distributions (Box Plot)")
axes[0].set_ylabel("Score")

# StudyHours — likely has outliers
sns.boxplot(data=df, y="StudyHours", ax=axes[1], width=0.4, color="lightblue")
axes[1].set_title("Study Hours Distribution")

plt.tight_layout()
plt.show()
````

### 8.2 Grouped Box Plot

Compare distributions **across categories** — this is where box plots truly shine in EDA.

````python
# Midterm scores by Major
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x="Major", y="Midterm", ax=axes[0], palette="Set2")
axes[0].set_title("Midterm by Major")
axes[0].tick_params(axis="x", rotation=30)

# Final scores by Gender
sns.boxplot(data=df, x="Gender", y="Final", ax=axes[1], palette="Set2")
axes[1].set_title("Final by Gender")

plt.tight_layout()
plt.show()
````

### 8.3 Violin Plot — Box Plot + KDE

A violin plot combines a box plot with a KDE on both sides — showing the **full distribution shape**, not just summary statistics.

````python
# Violin plot: Final scores by Major
fig, ax = plt.subplots(figsize=(10, 5))

sns.violinplot(data=df, x="Major", y="Final", ax=ax, palette="muted", inner="quartile")
ax.set_title("Final Score by Major (Violin Plot)")
ax.tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
````

## 9. Outlier Detection

Outliers are data points that are **far from the rest**. They can be:
- **Legitimate** (a genius student scoring 100) → keep them
- **Errors** (data entry mistake: age = 200) → fix or remove them

### Methods for Detecting Outliers

| Method | Rule | Best For |
|--------|------|----------|
| **IQR Method** | Below Q1−1.5×IQR or above Q3+1.5×IQR | Non-normal data |
| **Z-Score** | \|z\| > 3 (more than 3 std from mean) | Normal data |

### 9.1 IQR Method

The IQR method uses the "1.5×IQR rule" — any value outside the fences is flagged as an outlier.

````python
def find_outliers_iqr(series, multiplier=1.5):
    """Find outliers using the IQR method."""
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_fence = q1 - multiplier * iqr
    upper_fence = q3 + multiplier * iqr
    outliers = series[(series < lower_fence) | (series > upper_fence)]
    return outliers, lower_fence, upper_fence

# Find outliers in StudyHours
outliers, lower, upper = find_outliers_iqr(df["StudyHours"])
print(f"=== StudyHours Outliers (IQR Method) ===")
print(f"Lower fence: {lower:.1f}")
print(f"Upper fence: {upper:.1f}")
print(f"Outliers found: {len(outliers)}")
if len(outliers) > 0:
    print(f"Values: {sorted(outliers.values)}")

# Check all numeric columns
print("\n=== Outlier Count per Column ===")
for col in ["Midterm", "Final", "Homework", "StudyHours", "Attendance"]:
    out, _, _ = find_outliers_iqr(df[col].dropna())
    print(f"  {col:12s}: {len(out)} outliers")
````

### 9.2 Z-Score Method

The Z-score measures how many standard deviations a value is from the mean.  
$z = \frac{x - \bar{x}}{s}$  
Values with |z| > 3 are typically considered outliers.

````python
# Z-score method for outlier detection
study = df["StudyHours"].dropna()

# Calculate z-scores
z_scores = (study - study.mean()) / study.std()

# Or use scipy
z_scores_scipy = np.abs(stats.zscore(study))

# Find outliers (|z| > 3)
threshold = 3
outlier_mask = np.abs(z_scores) > threshold
print(f"=== Z-Score Outliers (|z| > {threshold}) ===")
print(f"Outliers found: {outlier_mask.sum()}")
print(f"Values: {study[outlier_mask].values}")
print(f"Z-scores: {z_scores[outlier_mask].round(2).values}")

# Also check with threshold = 2
outlier_mask_2 = np.abs(z_scores) > 2
print(f"\nWith |z| > 2: {outlier_mask_2.sum()} outliers")
````

### 9.3 Visualizing Outliers

A box plot with individual data points overlaid makes outliers immediately visible.

````python
# Visualize outliers with box + strip plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# StudyHours — box + strip
sns.boxplot(data=df, y="StudyHours", ax=axes[0], width=0.3, color="lightblue")
sns.stripplot(data=df, y="StudyHours", ax=axes[0], color="red", alpha=0.4, size=5)
axes[0].set_title("Study Hours: Box + Individual Points")

# Z-score distribution
z_all = np.abs(stats.zscore(df["StudyHours"].dropna()))
axes[1].hist(z_all, bins=20, color="steelblue", edgecolor="white")
axes[1].axvline(3, color="red", linestyle="--", label="z = 3")
axes[1].axvline(2, color="orange", linestyle="--", label="z = 2")
axes[1].set_title("Z-Score Distribution (StudyHours)")
axes[1].set_xlabel("|Z-score|")
axes[1].legend()

plt.tight_layout()
plt.show()
````

## 10. Correlation Analysis

Correlation measures the **linear relationship** between two numerical variables.

| Value | Interpretation |
|-------|----------------|
| **+1** | Perfect positive linear relationship |
| **+0.7 to +1** | Strong positive |
| **+0.3 to +0.7** | Moderate positive |
| **0 to +0.3** | Weak positive |
| **0** | No linear relationship |
| **Negative** | Same scale, inverse direction |

> **Warning**: Correlation ≠ Causation! Two variables can be correlated without one causing the other.

### 10.1 Correlation Matrix

Compute the Pearson correlation matrix for all numeric columns.

````python
# Correlation matrix (Pearson by default)
num_df = df[["Midterm", "Final", "Homework", "Attendance", "StudyHours"]]
corr_matrix = num_df.corr().round(3)
print("=== Pearson Correlation Matrix ===")
print(corr_matrix)
````

### 10.2 Heatmap — Visualizing Correlations

A heatmap makes it easy to spot strong and weak correlations at a glance.

````python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(
    corr_matrix,
    annot=True,          # Show numbers
    fmt=".2f",           # Two decimal places
    cmap="RdBu_r",       # Red=positive, Blue=negative
    center=0,            # Center colormap at 0
    vmin=-1, vmax=1,     # Full correlation range
    square=True,         # Square cells
    linewidths=0.5,
    ax=ax
)
ax.set_title("Correlation Heatmap", fontsize=14)
plt.tight_layout()
plt.show()
````

### 10.3 Scatter Plots & `pairplot()`

Scatter plots show the actual relationship between two variables. `pairplot()` creates a scatter matrix for all numeric columns.

````python
# Scatter plot: Midterm vs Final
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Midterm vs Final
sns.scatterplot(data=df, x="Midterm", y="Final", ax=axes[0], alpha=0.6)
sns.regplot(data=df, x="Midterm", y="Final", ax=axes[0],
            scatter=False, color="red", line_kws={"linewidth": 2})
r = df["Midterm"].corr(df["Final"])
axes[0].set_title(f"Midterm vs Final (r = {r:.3f})")

# StudyHours vs Final
sns.scatterplot(data=df, x="StudyHours", y="Final", ax=axes[1], alpha=0.6)
sns.regplot(data=df, x="StudyHours", y="Final", ax=axes[1],
            scatter=False, color="red", line_kws={"linewidth": 2})
r2 = df["StudyHours"].corr(df["Final"])
axes[1].set_title(f"StudyHours vs Final (r = {r2:.3f})")

plt.tight_layout()
plt.show()
````

Seaborn's `pairplot()` creates **all pairwise scatter plots** in one call — great for quick exploration.

````python
# Pairplot for key numeric variables, colored by Gender
sns.pairplot(
    df[["Midterm", "Final", "StudyHours", "Gender"]].dropna(),
    hue="Gender",
    diag_kind="kde",
    palette="Set1",
    plot_kws={"alpha": 0.5},
)
plt.suptitle("Pairplot: Exam Scores & Study Hours by Gender", y=1.02)
plt.show()
````

## 11. EDA Workflow: Putting It All Together

A systematic EDA follows a repeatable process:

```
1. Load Data → 2. Check Structure → 3. Handle Missing Values
↓
4. Univariate Analysis (each column alone)
↓
5. Bivariate Analysis (pairs of columns)
↓
6. Outlier Detection → 7. Insights & Questions
```

### EDA Checklist

| Step | What to Do | Tools |
|------|-----------|-------|
| Structure | `shape`, `dtypes`, `info()` | Pandas |
| Missing data | `isnull().sum()`, visualize with heatmap | Pandas, Seaborn |
| Summary stats | `describe()`, skew, kurtosis | Pandas, SciPy |
| Distributions | Histograms, KDE, box plots | Seaborn |
| Categories | `value_counts()`, bar charts, cross-tabs | Pandas, Seaborn |
| Relationships | Correlation matrix, scatter plots, pairplot | Pandas, Seaborn |
| Outliers | IQR method, z-scores | Custom functions |
| Group comparisons | Grouped box/bar plots, groupby stats | Pandas, Seaborn |

### 11.1 Missing Data Visualization

Visualize the pattern of missing values — random or systematic?

````python
# Missing value summary
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(1)
missing_df = pd.DataFrame({"Count": missing, "Percent": missing_pct})
print("=== Missing Values ===")
print(missing_df[missing_df["Count"] > 0])

# Visualize missing pattern as a heatmap
fig, ax = plt.subplots(figsize=(10, 4))
sns.heatmap(df.isnull().T, cbar=True, cmap="YlOrRd", yticklabels=True, ax=ax)
ax.set_title("Missing Data Pattern (yellow = missing)")
ax.set_xlabel("Row Index")
plt.tight_layout()
plt.show()
````

### 11.2 Distribution Overview — All Numeric Columns at Once

Plot histograms for all numeric columns in a single figure — fast visual scan.

````python
# Histograms for all numeric columns
num_columns = ["Age", "Midterm", "Final", "Homework", "Attendance", "StudyHours"]
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(num_columns):
    sns.histplot(df[col].dropna(), kde=True, ax=axes[i], color=sns.color_palette()[i])
    axes[i].set_title(f"{col} (skew={df[col].skew():.2f})")
    axes[i].axvline(df[col].mean(), color="red", linestyle="--", alpha=0.7)

plt.suptitle("Distribution of All Numeric Variables", fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
````

### 11.3 Group Comparison

Compare numeric variables across categories — do scores differ by Major or Gender?

````python
# GroupBy summary: performance by Major
major_summary = df.groupby("Major").agg(
    N=("Name", "count"),
    Midterm_Avg=("Midterm", "mean"),
    Final_Avg=("Final", "mean"),
    Study_Avg=("StudyHours", "mean"),
    Attend_Avg=("Attendance", "mean"),
).round(1)

print("=== Performance by Major ===")
print(major_summary)

# Grouped bar chart
fig, ax = plt.subplots(figsize=(10, 5))
major_summary[["Midterm_Avg", "Final_Avg"]].plot(kind="bar", ax=ax, colormap="Set2")
ax.set_title("Average Exam Scores by Major")
ax.set_ylabel("Score")
ax.set_xlabel("Major")
ax.legend(["Midterm", "Final"])
ax.tick_params(axis="x", rotation=30)
plt.tight_layout()
plt.show()
````

## 12. Practical Example: Complete EDA on a Real Dataset

Let's perform a full EDA pipeline on our student dataset, following the systematic workflow from Section 11.

**Step 1: Load and inspect the data.**

````python
# Step 1: Load data
df = pd.read_csv("students_eda.csv")

print(f"Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nData types:\n{df.dtypes}")
print(f"\n--- First 5 rows ---")
df.head()
````

**Step 2: Check for missing values and decide how to handle them.**

````python
# Step 2: Missing values
print("=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0])

# Fill with median (robust to skewness)
for col in ["Midterm", "Final", "Attendance"]:
    n_missing = df[col].isnull().sum()
    if n_missing > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"  Filled {col}: {n_missing} values → median={median_val:.1f}")

print(f"\nMissing after filling: {df.isnull().sum().sum()}")
````

**Step 3: Compute overall summary statistics.**

````python
# Step 3: Descriptive statistics
print("=== Descriptive Statistics ===")
print(df.describe().round(2))

# Skewness and kurtosis
print("\n=== Shape Statistics ===")
for col in ["Midterm", "Final", "Homework", "StudyHours"]:
    print(f"  {col:12s}  skew={df[col].skew():.2f}  kurtosis={df[col].kurtosis():.2f}")
````

**Step 4: Feature engineering — compute a weighted overall score and assign grades.**

````python
# Step 4: Feature engineering
df["Overall"] = (
    df["Midterm"] * 0.30 +
    df["Final"] * 0.50 +
    df["Homework"] * 0.20
).round(2)

# Assign letter grades
bins = [0, 60, 70, 80, 90, 100]
labels = ["F", "D", "C", "B", "A"]
df["Grade"] = pd.cut(df["Overall"], bins=bins, labels=labels, right=True)

print("=== Grade Distribution ===")
grade_counts = df["Grade"].value_counts().sort_index()
for grade, count in grade_counts.items():
    pct = count / len(df) * 100
    bar = "█" * count
    print(f"  {grade}: {count:>2} ({pct:5.1f}%) {bar}")

print(f"\nClass average: {df['Overall'].mean():.1f}")
print(f"Pass rate (≥60): {(df['Overall'] >= 60).mean()*100:.1f}%")
````

**Step 5: Univariate analysis — visualize all key distributions.**

````python
# Step 5: Univariate analysis
fig, axes = plt.subplots(2, 3, figsize=(16, 9))

# Row 1: Exam score distributions
for i, col in enumerate(["Midterm", "Final", "Homework"]):
    sns.histplot(df[col], kde=True, ax=axes[0, i], color=sns.color_palette()[i])
    axes[0, i].axvline(df[col].mean(), color="red", linestyle="--", label=f"Mean={df[col].mean():.1f}")
    axes[0, i].axvline(df[col].median(), color="orange", linestyle="--", label=f"Med={df[col].median():.1f}")
    axes[0, i].set_title(f"{col} Distribution")
    axes[0, i].legend(fontsize=8)

# Row 2: Other variables
sns.histplot(df["Overall"], kde=True, ax=axes[1, 0], color="purple")
axes[1, 0].set_title("Overall Score")

sns.histplot(df["StudyHours"], kde=True, ax=axes[1, 1], color="teal")
axes[1, 1].set_title("Study Hours/Week")

sns.countplot(data=df, x="Grade", ax=axes[1, 2],
              order=["A", "B", "C", "D", "F"], palette="RdYlGn_r")
axes[1, 2].set_title("Grade Distribution")

plt.suptitle("Univariate Analysis", fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
````

**Step 6: Bivariate analysis — relationships between variables.**

````python
# Step 6: Bivariate analysis — Correlation heatmap
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Correlation heatmap
corr = df[["Midterm", "Final", "Homework", "Attendance", "StudyHours", "Overall"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdBu_r", center=0,
            square=True, linewidths=0.5, ax=axes[0])
axes[0].set_title("Correlation Heatmap")

# Key scatter: StudyHours vs Overall
sns.scatterplot(data=df, x="StudyHours", y="Overall", hue="Grade",
                palette="RdYlGn_r", ax=axes[1], alpha=0.7)
sns.regplot(data=df, x="StudyHours", y="Overall", scatter=False,
            color="gray", ax=axes[1], line_kws={"linewidth": 2})
r = df["StudyHours"].corr(df["Overall"])
axes[1].set_title(f"Study Hours vs Overall Score (r={r:.3f})")

plt.tight_layout()
plt.show()
````

**Step 7: Group comparisons — how do different groups perform?**

````python
# Step 7: Group comparisons
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Overall score by Major
sns.boxplot(data=df, x="Major", y="Overall", ax=axes[0], palette="Set2")
axes[0].set_title("Overall Score by Major")
axes[0].tick_params(axis="x", rotation=30)

# Overall score by Gender
sns.violinplot(data=df, x="Gender", y="Overall", ax=axes[1],
               palette="Set2", inner="quartile")
axes[1].set_title("Overall Score by Gender")

# Grade distribution by Major
grade_by_major = pd.crosstab(df["Major"], df["Grade"], normalize="index") * 100
grade_by_major[["A", "B", "C", "D", "F"]].plot(
    kind="bar", stacked=True, ax=axes[2], colormap="RdYlGn_r")
axes[2].set_title("Grade Distribution by Major (%)")
axes[2].set_ylabel("Percentage")
axes[2].tick_params(axis="x", rotation=30)
axes[2].legend(title="Grade", bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
````

**Step 8: Outlier check and final statistical summary.**

````python
# Step 8: Outlier summary
print("=== Outlier Detection (IQR Method) ===")
for col in ["Midterm", "Final", "Homework", "StudyHours", "Attendance", "Overall"]:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"  {col:12s}  Fences=[{lower:.1f}, {upper:.1f}]  Outliers={n_out}")

# Comprehensive group summary
print("\n=== Final Summary by Major ===")
summary = df.groupby("Major").agg(
    N=("Name", "count"),
    Avg_Overall=("Overall", "mean"),
    Std_Overall=("Overall", "std"),
    Avg_Study=("StudyHours", "mean"),
    Avg_Attend=("Attendance", "mean"),
).round(1)
print(summary)
````

**Step 9: Key findings — summarize what the EDA revealed.**

````python
# Step 9: Key findings
print("=" * 60)
print("EDA KEY FINDINGS")
print("=" * 60)

print(f"\n1. DATASET: {len(df)} students, {df['Major'].nunique()} majors")
print(f"   - Most common major: {df['Major'].mode()[0]}")
print(f"   - Gender split: {df['Gender'].value_counts().to_dict()}")

print(f"\n2. PERFORMANCE:")
print(f"   - Class average: {df['Overall'].mean():.1f} ± {df['Overall'].std():.1f}")
print(f"   - Range: [{df['Overall'].min():.1f}, {df['Overall'].max():.1f}]")
print(f"   - Pass rate: {(df['Overall'] >= 60).mean()*100:.1f}%")

best_major = df.groupby('Major')['Overall'].mean().idxmax()
print(f"\n3. TOP PERFORMING MAJOR: {best_major}")
print(f"   - Average: {df.groupby('Major')['Overall'].mean()[best_major]:.1f}")

top_corr = corr["Overall"].drop("Overall").abs().idxmax()
print(f"\n4. STRONGEST PREDICTOR of Overall: {top_corr}")
print(f"   - Correlation: r = {corr['Overall'][top_corr]:.3f}")

print(f"\n5. STUDY HOURS: Mean={df['StudyHours'].mean():.1f}, ")
print(f"   Median={df['StudyHours'].median():.1f} (right-skewed: skew={df['StudyHours'].skew():.2f})")

print("\n" + "=" * 60)
````
