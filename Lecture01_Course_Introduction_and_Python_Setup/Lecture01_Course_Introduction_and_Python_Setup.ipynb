## 5. Why Python? (And Not R, MATLAB, or Others?)

This is a statistics course. You might wonder: why Python and not R, MATLAB, SAS, or SPSS?

Let's compare honestly.

### 5.1 Python vs R — The Big Debate

| Dimension | Python | R |
|-----------|--------|---|
| **Born for** | General-purpose programming | Statistical computing |
| **Syntax** | Clean, beginner-friendly | Quirky (e.g., `<-` for assignment) |
| **Statistics** | Great (scipy, statsmodels, pingouin) | Excellent (built-in, CRAN packages) |
| **Machine Learning** | Dominant (scikit-learn, PyTorch, TensorFlow) | Good (caret, tidymodels) but smaller ecosystem |
| **Deep Learning / AI** | Industry standard (PyTorch, HuggingFace) | Very limited |
| **Visualization** | Matplotlib, Seaborn, Plotly | ggplot2 (beautiful, but R-only) |
| **Web Scraping** | Excellent (requests, BeautifulSoup, Selenium) | Limited |
| **Job Market** | #1 in tech, data science, AI | Strong in academia, biostatistics |
| **Community** | Massive (Stack Overflow, GitHub) | Large but more academic |
| **Learning Curve** | Easy first language | Steeper for non-statisticians |

### 5.2 Python vs MATLAB / SAS / SPSS

| Tool | Strengths | Why We Don't Use It |
|------|-----------|---------------------|
| **MATLAB** | Engineering, signal processing, simulation | Expensive license ($$$), weak in ML/AI |
| **SAS** | Enterprise analytics, clinical trials | Expensive, declining market share, closed ecosystem |
| **SPSS** | Point-and-click statistics, social science | Not programmable, limited to traditional stats |
| **Stata** | Econometrics, panel data | Niche, expensive, no ML/AI support |
| **Excel** | Everyone knows it | Not scalable, error-prone, no reproducibility |

### 5.3 The Real Reason: Python Is the Language of AI

```
                    2015            2020            2025
                    ┌───┐           ┌───┐           ┌───┐
Statistics:         │ R │           │R/Py│           │ Py│  ← Python catching up
                    └───┘           └───┘           └───┘
Machine Learning:   ┌───┐           ┌───┐           ┌───┐
                    │R/Py│          │ Py│            │ Py│  ← Python dominates
                    └───┘           └───┘           └───┘
Deep Learning:      ┌───┐           ┌───┐           ┌───┐
                    │ Py│           │ Py│            │ Py│  ← Always Python
                    └───┘           └───┘           └───┘
LLMs / GenAI:         —              —             ┌───┐
                                                   │ Py│  ← Python only
                                                   └───┘
```

> **Bottom line:** If you learn Python, you can do statistics AND machine learning AND AI.  
> If you learn R, you can do statistics well, but you'll need to learn Python later anyway.  
> **We choose Python because it gives you the most return on your learning investment.**

### 5.5 "But My Professor Uses R..."

That's fine! The statistical **concepts** are the same regardless of language.

| Concept | Python | R |
|---------|--------|---|
| t-test | `scipy.stats.ttest_ind(a, b)` | `t.test(a, b)` |
| Linear regression | `statsmodels.OLS(y, X).fit()` | `lm(y ~ x, data)` |
| Histogram | `plt.hist(data)` | `hist(data)` |

Once you understand the statistics, switching between languages is easy.  
But the **programming skills, ML skills, and AI skills** you learn in Python are not easily replicated in R.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random exam scores — a Normal Distribution
np.random.seed(42)
scores = np.random.normal(loc=75, scale=10, size=1000)

plt.figure(figsize=(10, 5))
plt.hist(scores, bins=30, color="#3498db", edgecolor="white", alpha=0.8)
plt.axvline(x=np.mean(scores), color="red", linestyle="--", linewidth=2,
            label=f"Mean = {np.mean(scores):.1f}")
plt.title("Distribution of 1000 Exam Scores", fontsize=16)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Mean: {np.mean(scores):.1f}  |  Std: {np.std(scores):.1f}  |  Min: {np.min(scores):.1f}  |  Max: {np.max(scores):.1f}")
````

> This is a **Normal Distribution** — the bell curve, the most important distribution in statistics.  
> We generated it with just 2 lines of Python. We'll study it in depth in **Week 6**.

````python
# Does studying more lead to higher scores?
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 50)
exam_score = 50 + 4 * study_hours + np.random.normal(0, 5, 50)

plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_score, color="#3498db", alpha=0.7, edgecolors="white", s=80)

# Trend line
z = np.polyfit(study_hours, exam_score, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 10, 100)
plt.plot(x_line, p(x_line), color="#e74c3c", linewidth=2, linestyle="--",
         label=f"Trend: score = {z[0]:.1f} × hours + {z[1]:.1f}")

plt.title("Study Hours vs Exam Score", fontsize=16)
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

> **Think about it:**
> - Is there a relationship between study hours and exam scores?
> - Is this **correlation** or **causation**?
> - How confident can we be in this trend line?
>
> These are exactly the questions you'll answer in **Week 9: Regression**.
