## 1. Why Pandas?

### NumPy Limitations

NumPy arrays are powerful but limited for real data analysis:

| Problem | NumPy | Pandas |
|---------|-------|--------|
| Column names | No — just index numbers | Yes — named columns |
| Mixed data types | No — all same dtype | Yes — each column can differ |
| Missing values | Limited (`np.nan`) | Rich support (`NaN`, `dropna`, `fillna`) |
| Data alignment | Manual | Automatic by label |
| File I/O | Basic | One-line CSV, Excel, SQL, JSON... |
| GroupBy / Pivot | Manual loops | Built-in methods |

### The Two Core Objects

| Object | Dimensions | Analogy |
|--------|------------|--------|
| **Series** | 1D | A single column in a spreadsheet |
| **DataFrame** | 2D | An entire spreadsheet / SQL table |

## 2. Series — 1D Labeled Data

A Series is like a NumPy array, but each element has a **label** (index).  
Think of it as a single column in a spreadsheet, with row labels on the left.

### 2.1 Creating a Series

Create a Series from a **list** — Pandas auto-generates integer indices (0, 1, 2, ...).

````python
# From a list
scores = pd.Series([78, 92, 85, 67, 95])
print(scores)
print(f"\ntype: {type(scores)}")
print(f"dtype: {scores.dtype}")
````

Create a Series with **custom labels** — much more informative for real data.

````python
# With custom index
scores = pd.Series(
    [78, 92, 85, 67, 95],
    index=["Alice", "Bob", "Charlie", "Diana", "Eve"],
    name="Exam Score"
)
print(scores)
````

Create a Series from a **dictionary** — keys become the index labels automatically.

````python
# From a dictionary
gpa = pd.Series({
    "Alice": 3.8,
    "Bob": 3.2,
    "Charlie": 3.6,
    "Diana": 3.9,
})
print(gpa)
````

### 2.2 Series Indexing

You can access elements by **label** (like a dictionary) or by **position** (like a list).

````python
scores = pd.Series(
    [78, 92, 85, 67, 95],
    index=["Alice", "Bob", "Charlie", "Diana", "Eve"]
)

# By label
print(f"By label: scores['Bob']    = {scores['Bob']}")
print(f"By label: scores[['Alice','Eve']] =\n{scores[['Alice','Eve']]}")

# By position
print(f"\nBy position: scores.iloc[0]  = {scores.iloc[0]}")
print(f"By position: scores.iloc[-1] = {scores.iloc[-1]}")
````

### 2.3 Series Operations

Like NumPy, operations are **vectorized** — they apply to every element automatically.

````python
scores = pd.Series([78, 92, 85, 67, 95],
                    index=["Alice", "Bob", "Charlie", "Diana", "Eve"])

# Vectorized operations
print(f"Curved scores (add 5):\n{scores + 5}\n")

# Boolean filtering
print(f"Scores ≥ 85:\n{scores[scores >= 85]}\n")

# Built-in statistics
print(f"Mean:   {scores.mean():.1f}")
print(f"Std:    {scores.std():.1f}")
print(f"Median: {scores.median():.1f}")
print(f"Min:    {scores.min()}")
print(f"Max:    {scores.max()}")
````

## 3. DataFrame — 2D Tabular Data

A DataFrame is the star of Pandas — a 2D table with **labeled rows and columns**.  
Think of it as:
- An Excel spreadsheet
- A SQL table
- A collection of Series that share the same index

### 3.1 Creating a DataFrame

**From a dictionary of lists** — each key becomes a column name, each list becomes column values.

````python
# From a dictionary of lists
df = pd.DataFrame({
    "Name":   ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age":    [22, 21, 23, 22, 20],
    "Major":  ["Data Science", "Statistics", "Math", "Data Science", "CS"],
    "GPA":    [3.8, 3.2, 3.6, 3.9, 3.5],
    "Score":  [92, 78, 85, 95, 88],
})
df
````

**From a list of dictionaries** — each dictionary becomes a row.

````python
# From a list of dictionaries
records = [
    {"Name": "Alice", "Age": 22, "Score": 92},
    {"Name": "Bob",   "Age": 21, "Score": 78},
    {"Name": "Charlie", "Age": 23, "Score": 85},
]
df2 = pd.DataFrame(records)
df2
````

**From a NumPy array** — provide column names manually.

````python
# From a NumPy array
rng = np.random.default_rng(42)
data = rng.integers(60, 100, size=(5, 3))

df3 = pd.DataFrame(data, columns=["Midterm", "Final", "Project"])
df3
````

### 4.3 Writing and Reading Excel

Excel is the most common data format in business and academia. Pandas handles it natively.

> **Note:** You may need to install `openpyxl` first: `pip install openpyxl`

**Write a DataFrame to Excel** — use `to_excel()`. Each DataFrame becomes one sheet.

````python
# Write DataFrame to a single-sheet Excel file
df = pd.read_csv("students.csv")
df.to_excel("students.xlsx", index=False, sheet_name="Sheet1")
print("Saved students.xlsx")
print(df.head())
````

**Read an Excel file** — `pd.read_excel()` works just like `read_csv()`. By default it reads the first sheet.

````python
# Read the Excel file back
df_excel = pd.read_excel("students.xlsx")
print(f"Shape: {df_excel.shape}")
df_excel.head()
````

**Write multiple sheets** to one Excel file — use `ExcelWriter` as a context manager. This is like having multiple tabs in one workbook.

````python
# Write multiple sheets to one Excel file
df = pd.read_csv("students.csv")

# Split data by major
ds_students = df[df["Major"] == "DS"]
stats_students = df[df["Major"] == "Stats"]
math_students = df[df["Major"] == "Math"]

# Write each group to a separate sheet using ExcelWriter
with pd.ExcelWriter("students_by_major.xlsx") as writer:
    df.to_excel(writer, sheet_name="All Students", index=False)
    ds_students.to_excel(writer, sheet_name="Data Science", index=False)
    stats_students.to_excel(writer, sheet_name="Statistics", index=False)
    math_students.to_excel(writer, sheet_name="Mathematics", index=False)

print("Saved students_by_major.xlsx with 4 sheets")
````

**Read a specific sheet** — use `sheet_name` to choose which tab to read. You can also read all sheets at once into a dictionary.

````python
# Read a specific sheet by name
df_ds = pd.read_excel("students_by_major.xlsx", sheet_name="Data Science")
print("--- Data Science sheet ---")
print(df_ds)

# Read a specific sheet by index (0 = first sheet)
df_first = pd.read_excel("students_by_major.xlsx", sheet_name=0)
print(f"\n--- First sheet (All Students): {len(df_first)} rows ---")

# Read ALL sheets at once → returns a dict of DataFrames
all_sheets = pd.read_excel("students_by_major.xlsx", sheet_name=None)
print(f"\nAll sheet names: {list(all_sheets.keys())}")
for name, sheet_df in all_sheets.items():
    print(f"  {name}: {sheet_df.shape}")
````

### 4.4 Useful `read_csv()` / `read_excel()` / `read_json()` Parameters

These three functions share many common parameters (`header`, `usecols`, `skiprows`, `nrows`, `dtype`, etc.), 
but each also has format-specific options. Let's see them **in action** with real data.

#### `read_csv()` — Common Parameters in Action

````python
# ---- read_csv(): Common Parameters Demo ----

# 1) Basic read — row 0 is used as column header by default
df_basic = pd.read_csv("students.csv")
print("=== 1. Basic read ===")
print(df_basic)
print()
````

````python
# 2) usecols — read only specific columns (saves memory on large files)
df_cols = pd.read_csv("students.csv", usecols=["Name", "Major", "Final"])
print("=== 2. usecols: only Name, Major, Final ===")
print(df_cols)
print()
````

````python
# 3) index_col — use a column as the row index
df_idx = pd.read_csv("students.csv", index_col="Name")
print("=== 3. index_col: use Name as row index ===")
print(df_idx)
print()
````

````python
# 4) nrows — read only the first N rows (great for previewing large files)
df_head = pd.read_csv("students.csv", nrows=3)
print("=== 4. nrows=3: only first 3 rows ===")
print(df_head)
print()
````

````python
# 5) skiprows — skip specific rows (accepts int or list)
df_skip = pd.read_csv("students.csv", skiprows=[1, 3])  # skip row 1 (Alice) and row 3 (Charlie)
print("=== 5. skiprows=[1,3]: skip Alice and Charlie ===")
print(df_skip)
print()
````

````python
# 6) dtype — force column data types
#    e.g., keep Age as string instead of int
df_dtype = pd.read_csv("students.csv", dtype={"Age": str, "Midterm": float})
print("=== 6. dtype: Age->str, Midterm->float ===")
print(df_dtype.dtypes)
print()
print(df_dtype.head(3))
````

````python
# 7) na_values — specify which strings should be treated as NaN
#    Let's create a small CSV with messy missing-value markers
import io

csv_with_na = """Name,Age,Major,Midterm,Final
Alice,22,DS,85,90
Bob,N/A,Stats,--,82
Charlie,23,Math,92,missing"""

df_na = pd.read_csv(io.StringIO(csv_with_na), na_values=["N/A", "--", "missing"])
print("=== 7. na_values: treat N/A, --, missing as NaN ===")
print(df_na)
print(f"\nMissing value count per column:\n{df_na.isna().sum()}")
````

#### `read_json()` — Reading JSON Data

JSON (JavaScript Object Notation) is the most common data format for Web APIs. Pandas can read and write JSON directly.

````python
# ---- First, create a JSON file from our CSV data ----

df = pd.read_csv("students.csv")

# orient="records": each row becomes a dict — the most common API format
# [{"Name":"Alice","Age":22,...}, {"Name":"Bob",...}]
df.to_json("students.json", orient="records", indent=2, force_ascii=False)

print("students.json created! Preview:")
with open("students.json") as f:
    print(f.read())
````

````python
# ---- read_json(): Common Parameters Demo ----

# 1) Basic read — orient must match how the file was written
df_json = pd.read_json("students.json", orient="records")
print("=== 1. Basic JSON read (orient='records') ===")
print(df_json)
print()
````

````python
# 2) orient — different JSON structures (JSON-specific!)
#    'records' : [{col:val, ...}, ...]       ← most common (API responses)
#    'columns' : {col: {idx:val, ...}, ...}  ← organized by column
#    'index'   : {idx: {col:val, ...}, ...}  ← organized by row index
#    'split'   : {columns:[...], data:[...]}  ← columns and data separated

df_small = df.head(3)

for orient in ["records", "columns", "index", "split"]:
    json_str = df_small.to_json(orient=orient, indent=2)
    print(f'=== orient="{orient}" ===')
    print(json_str[:200])  # show first 200 chars
    print()
````

````python
# 3) dtype — force column types (same as read_csv / read_excel)
df_json_dtype = pd.read_json("students.json", orient="records", dtype={"Age": str})
print("=== 3. dtype: Age->str ===")
print(df_json_dtype.dtypes)
print()
print(df_json_dtype.head(3))
````

````python
# 4) Read JSON from a string (no file needed)
#    Useful when working with Web API responses
import io

json_string = '[{"city":"Beijing","temp":32},{"city":"Shanghai","temp":35},{"city":"Guangzhou","temp":33}]'

df_api = pd.read_json(io.StringIO(json_string), orient="records")
print("=== 4. Read from JSON string (simulating API response) ===")
print(df_api)
````

## 6. Indexing & Selection

Selecting data is the most common operation in Pandas. There are several methods:

| Method | Selects by | Example |
|--------|-----------|--------|
| `df["col"]` | Column name | `df["Age"]` |
| `df[["A","B"]]` | Multiple columns | `df[["Name","Score"]]` |
| `df.loc[row, col]` | Labels | `df.loc[0, "Name"]` |
| `df.iloc[row, col]` | Integer positions | `df.iloc[0, 0]` |
| `df[condition]` | Boolean mask | `df[df["Age"] > 21]` |

### 6.1 Selecting Columns

Select a **single column** (returns a Series) or **multiple columns** (returns a DataFrame).

````python
# Single column → Series
print(type(df["Name"]))
print(df["Name"])

print("\n--- Multiple columns → DataFrame ---")
print(df[["Name", "Midterm", "Final"]])
````

### 6.2 `loc` — Label-Based Selection

Use `loc[row_label, col_label]` to select by **labels**. Both row and column can be single labels, lists, or slices.

````python
# loc: label-based
print("Single element:")
print(f"  df.loc[0, 'Name'] = {df.loc[0, 'Name']}")

print("\nSingle row:")
print(df.loc[0])

print("\nRows 0-2, specific columns:")
print(df.loc[0:2, ["Name", "Midterm", "Final"]])
````

### 6.3 `iloc` — Integer Position-Based Selection

Use `iloc[row_pos, col_pos]` to select by **integer positions**. Works like NumPy indexing.

````python
# iloc: position-based
print(f"First row, first col: {df.iloc[0, 0]}")
print(f"Last row:  {df.iloc[-1].to_dict()}")

print("\nFirst 3 rows, columns 0 and 3:")
print(df.iloc[:3, [0, 3]])
````

### 6.4 Boolean Indexing (Filtering)

The most powerful selection method — use **conditions** to filter rows. This is how you'll select data in practice.

````python
df = pd.read_csv("students.csv")

# Single condition
print("--- Students with Midterm ≥ 85 ---")
print(df[df["Midterm"] >= 85])

# Multiple conditions (use & for AND, | for OR, ~ for NOT)
print("\n--- DS majors with Final ≥ 85 ---")
print(df[(df["Major"] == "DS") & (df["Final"] >= 85)])
````

The `isin()` method checks if values are in a given list — great for filtering by multiple categories.

````python
# isin() for multiple values
print("--- DS or Stats majors ---")
print(df[df["Major"].isin(["DS", "Stats"])])
````

The `query()` method lets you write conditions as a **string** — more readable for complex filters.

````python
# query() — more readable for complex filters
print("--- query: Age >= 22 and Midterm >= 85 ---")
print(df.query("Age >= 22 and Midterm >= 85"))
````

### 6.5 `between()` — Range Filtering

`between(low, high)` is a shortcut for `(col >= low) & (col <= high)` — cleaner and more readable.

````python
# between() — inclusive range filter
print("--- Midterm between 80 and 90 ---")
print(df[df["Midterm"].between(80, 90)])

# Combine with other conditions
print("\n--- Age 21-22 AND Midterm 80-95 ---")
mask = df["Age"].between(21, 22) & df["Midterm"].between(80, 95)
print(df[mask][["Name", "Age", "Midterm"]])
````

## 7. Data Manipulation

Transforming, adding, and removing columns is how you prepare data for analysis.

### 7.1 Adding New Columns

Add a calculated column — this is **vectorized**, so it runs on the entire column at once.

````python
df = pd.read_csv("students.csv")

# Calculated column: weighted average
df["Average"] = df["Midterm"] * 0.4 + df["Final"] * 0.6

# Conditional column using np.select()
conditions = [
    df["Average"] >= 90,
    df["Average"] >= 80,
    df["Average"] >= 70,
    df["Average"] >= 60,
]
grades = ["A", "B", "C", "D"]
df["Grade"] = np.select(conditions, grades, default="F")

print(df[["Name", "Midterm", "Final", "Average", "Grade"]])
````

### 7.2 Renaming Columns

````python
# Rename specific columns
df_renamed = df.rename(columns={"Midterm": "Mid", "Final": "Fin"})
print(f"Columns: {list(df_renamed.columns)}")

# Lowercase all columns
df_lower = df.rename(columns=str.lower)
print(f"Lowercase: {list(df_lower.columns)}")
````

### 7.3 Dropping Columns and Rows

````python
# Drop columns
df_slim = df.drop(columns=["Grade"])
print(f"After dropping 'Grade': {list(df_slim.columns)}")

# Drop rows by index
df_fewer = df.drop(index=[0, 1])
print(f"\nAfter dropping rows 0,1: {len(df_fewer)} rows remaining")
print(df_fewer.head())
````

### 7.4 Sorting

Sort by one or more columns — use `ascending=False` for descending order.

````python
# Sort by single column
print("--- Sorted by Average (descending) ---")
print(df.sort_values("Average", ascending=False)[["Name", "Average", "Grade"]])

# Sort by multiple columns
print("\n--- Sorted by Grade (asc), then Average (desc) ---")
print(df.sort_values(["Grade", "Average"], ascending=[True, False])[["Name", "Average", "Grade"]])
````

### 7.5 `nlargest()` and `nsmallest()` — Quick Top/Bottom N

````python
# Top 3 students by average
print("--- Top 3 ---")
print(df.nlargest(3, "Average")[["Name", "Average", "Grade"]])

# Bottom 3 students by average
print("\n--- Bottom 3 ---")
print(df.nsmallest(3, "Average")[["Name", "Average", "Grade"]])
````

### 7.6 String Methods (`.str`)

Pandas provides a `.str` accessor for **vectorized string operations** on text columns — no loops needed.

````python
df = pd.read_csv("students.csv")

# Convert to uppercase / lowercase
print("--- Upper case ---")
print(df["Name"].str.upper())

# Check if string contains a substring
print("\n--- Names containing 'a' (case-insensitive) ---")
print(df[df["Name"].str.lower().str.contains("a")]["Name"])

# Get string length
print("\n--- Name lengths ---")
df["Name_Len"] = df["Name"].str.len()
print(df[["Name", "Name_Len"]])

# Slice strings (first 3 characters)
print("\n--- First 3 chars ---")
print(df["Name"].str[:3])
````

### 7.7 `replace()` — Replacing Values

Replace specific values in a column or across the entire DataFrame.

````python
df = pd.read_csv("students.csv")

# Replace a single value
df_replaced = df.copy()
df_replaced["Major"] = df_replaced["Major"].replace("DS", "Data Science")
print("--- Replace 'DS' with 'Data Science' ---")
print(df_replaced["Major"].value_counts())

# Replace multiple values using a dictionary
df_replaced["Major"] = df_replaced["Major"].replace({
    "Stats": "Statistics",
    "Math": "Mathematics",
})
print("\n--- Replace multiple values ---")
print(df_replaced["Major"].value_counts())
````

### 7.8 `astype()` — Changing Data Types

Convert columns to a different data type. Common use cases: convert strings to numbers, integers to categories.

````python
df = pd.read_csv("students.csv")

# Check current types
print("--- Original dtypes ---")
print(df.dtypes)

# Convert Age to float
df["Age_float"] = df["Age"].astype(float)
print(f"\nAge as float: {df['Age_float'].dtype}")

# Convert Major to category type (saves memory for large datasets)
df["Major_cat"] = df["Major"].astype("category")
print(f"Major as category: {df['Major_cat'].dtype}")
print(f"Categories: {df['Major_cat'].cat.categories.tolist()}")

# Convert numeric to string
df["Midterm_str"] = df["Midterm"].astype(str)
print(f"\nMidterm as string: {df['Midterm_str'].dtype}")
print(f"Example: '{df['Midterm_str'].iloc[0]}' (type: {type(df['Midterm_str'].iloc[0]).__name__})")
````

## 8. Handling Missing Data

Real-world data almost always has **missing values**. Pandas represents them as `NaN` (Not a Number).  
Handling missing data correctly is crucial — ignoring it can lead to wrong conclusions.

### 8.1 Creating Data with Missing Values

````python
# Create DataFrame with missing values
df_missing = pd.DataFrame({
    "Name":    ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age":     [22, np.nan, 23, 22, 20],
    "Midterm": [85, 78, np.nan, 88, 76],
    "Final":   [90, 82, 88, np.nan, 80],
})
print(df_missing)
````

### 8.2 Detecting Missing Values

`isnull()` returns a boolean DataFrame — `True` where data is missing.

````python
# Detect missing values
print("--- isnull() ---")
print(df_missing.isnull())

# Count missing per column
print("\n--- Missing count per column ---")
print(df_missing.isnull().sum())

# Total missing
print(f"\nTotal missing values: {df_missing.isnull().sum().sum()}")
````

### 8.3 Dropping Missing Values

`dropna()` removes rows (or columns) that contain missing values. Use `how` and `subset` to control the behavior.

````python
# Drop rows with ANY missing value
print("--- dropna() ---")
print(df_missing.dropna())

# Drop only rows where ALL values are missing
print("\n--- dropna(how='all') ---")
print(df_missing.dropna(how="all"))

# Drop rows where specific columns are missing
print("\n--- dropna(subset=['Final']) ---")
print(df_missing.dropna(subset=["Final"]))
````

### 8.4 Filling Missing Values

Instead of dropping data, you can **fill** missing values with a replacement — mean, median, or a custom value.

````python
# Fill with a fixed value
print("--- Fill with 0 ---")
print(df_missing.fillna(0))
````

**Fill with the column mean** — a common strategy for numeric columns.

````python
# Fill with column mean (only for numeric columns)
df_filled = df_missing.copy()
numeric_cols = df_filled.select_dtypes(include="number").columns
df_filled[numeric_cols] = df_filled[numeric_cols].fillna(df_filled[numeric_cols].mean())

print("--- Fill with column mean ---")
print(df_filled)
````

**Forward fill (`ffill`)** and **backward fill (`bfill`)** — useful for time series or ordered data.

````python
# Forward fill (use previous value)
print("--- Forward fill ---")
print(df_missing.ffill())

# Backward fill (use next value)
print("\n--- Backward fill ---")
print(df_missing.bfill())
````

## 9. GroupBy — Split-Apply-Combine

GroupBy is one of Pandas' most powerful operations. It follows a three-step pattern:

1. **Split** — divide the data into groups based on some criteria
2. **Apply** — apply a function to each group independently
3. **Combine** — merge the results back together

```
df.groupby("column").aggregation()
```

### 9.1 Basic GroupBy

Group students by Major and compute the mean of each numeric column.

````python
df = pd.read_csv("students.csv")
df["Average"] = df["Midterm"] * 0.4 + df["Final"] * 0.6

# Group by Major → compute mean
print("--- Mean by Major ---")
print(df.groupby("Major")[["Midterm", "Final", "Average"]].mean().round(1))
````

### 9.2 Multiple Aggregations with `agg()`

Apply **different functions** to different columns, or multiple functions to the same column.

````python
# Multiple aggregation functions
result = df.groupby("Major").agg(
    Count=("Name", "count"),
    Avg_Midterm=("Midterm", "mean"),
    Avg_Final=("Final", "mean"),
    Max_Average=("Average", "max"),
    Min_Average=("Average", "min"),
).round(1)

print(result)
````

### 9.3 Group by Multiple Columns

````python
# Group by Major AND Age
print(df.groupby(["Major", "Age"])["Average"].mean().round(1))
````

### 9.4 Pivot Table

`pivot_table()` creates a spreadsheet-style summary table — like Excel's PivotTable. It combines groupby + reshape in one step.

````python
df = pd.read_csv("students.csv")
df["Average"] = df["Midterm"] * 0.4 + df["Final"] * 0.6

# Simple pivot table: average score by Major
pivot1 = df.pivot_table(
    values="Average",     # What to summarize
    index="Major",        # Row labels
    aggfunc="mean"        # How to aggregate (default is mean)
).round(1)
print("--- Average by Major ---")
print(pivot1)

# Pivot table with multiple aggregations
pivot2 = df.pivot_table(
    values=["Midterm", "Final"],
    index="Major",
    aggfunc=["mean", "max"]
).round(1)
print("\n--- Mean and Max by Major ---")
print(pivot2)

# Pivot table with row AND column grouping + margins (totals)
pivot3 = df.pivot_table(
    values="Average",
    index="Major",
    columns="Age",
    aggfunc="mean",
    margins=True,          # Add row/column totals
    margins_name="Overall"
).round(1)
print("\n--- Average by Major × Age (with totals) ---")
print(pivot3)
````

### 9.5 `size()` and `value_counts()`

Count the number of rows in each group.

````python
# Count per group
print("--- Group sizes ---")
print(df.groupby("Major").size())

# Same thing with value_counts()
print("\n--- value_counts ---")
print(df["Major"].value_counts())
````

## 10. Merge & Concatenation

Combining data from multiple sources is a daily task in data analysis.

### 10.1 Concatenation with `pd.concat()`

Stack DataFrames **vertically** (add rows) or **horizontally** (add columns).

**Vertical concatenation** — combine two DataFrames with the same columns (like adding more rows).

````python
# Two classes with the same columns
class_a = pd.DataFrame({
    "Name":  ["Alice", "Bob"],
    "Score": [92, 78],
})

class_b = pd.DataFrame({
    "Name":  ["Charlie", "Diana"],
    "Score": [85, 95],
})

# Vertical stack (default axis=0)
combined = pd.concat([class_a, class_b], ignore_index=True)
print("--- Vertical concat ---")
print(combined)
````

**Horizontal concatenation** — add new columns side by side.

````python
# Horizontal stack (axis=1)
names = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie"]})
scores = pd.DataFrame({"Midterm": [85, 78, 92], "Final": [90, 82, 88]})

combined = pd.concat([names, scores], axis=1)
print("--- Horizontal concat ---")
print(combined)
````

### 10.2 Merge (SQL-style Join)

`pd.merge()` combines DataFrames based on a **shared key column** — just like SQL JOIN.

| Join Type | Keeps |
|-----------|-------|
| `inner` | Only matching rows from both (default) |
| `left` | All rows from left + matches from right |
| `right` | All rows from right + matches from left |
| `outer` | All rows from both |

Let's create two separate tables and merge them on a shared key.

````python
# Table 1: Student info
students = pd.DataFrame({
    "ID":    [1, 2, 3, 4, 5],
    "Name":  ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Major": ["DS", "Stats", "Math", "DS", "CS"],
})

# Table 2: Exam scores
exams = pd.DataFrame({
    "ID":      [1, 2, 3, 4, 6],
    "Midterm": [85, 78, 92, 88, 70],
    "Final":   [90, 82, 88, 95, 75],
})

print("--- students ---")
print(students)
print("\n--- exams ---")
print(exams)
````

Notice: Student ID 5 (Eve) has no exam scores, and ID 6 has scores but no student info. The join type determines what happens.

````python
# Inner join: only IDs present in BOTH tables
inner = pd.merge(students, exams, on="ID", how="inner")
print("--- Inner join ---")
print(inner)

# Left join: keep ALL students, fill missing exams with NaN
left = pd.merge(students, exams, on="ID", how="left")
print("\n--- Left join ---")
print(left)

# Outer join: keep ALL rows from both tables
outer = pd.merge(students, exams, on="ID", how="outer")
print("\n--- Outer join ---")
print(outer)
````

## 11. Apply & Lambda

When built-in methods aren't enough, `apply()` lets you run **any function** on each row or column.

### 11.1 Apply to a Column (Series)

Apply a custom function to every value in a column — like `map()` for Series.

````python
df = pd.read_csv("students.csv")
df["Average"] = df["Midterm"] * 0.4 + df["Final"] * 0.6

# Custom function
def get_grade(score):
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

# Apply to a single column
df["Grade"] = df["Average"].apply(get_grade)
print(df[["Name", "Average", "Grade"]])
````

Using a **lambda function** for quick, one-line transformations.

````python
# Lambda: quick one-liner
df["Name_upper"] = df["Name"].apply(lambda x: x.upper())
df["Pass"] = df["Average"].apply(lambda x: "Yes" if x >= 60 else "No")

print(df[["Name", "Name_upper", "Average", "Pass"]])
````

### 11.2 Apply to Rows (`axis=1`)

Apply a function to **each row** — the function receives the entire row as a Series.

````python
# Apply to each row
def student_summary(row):
    return f"{row['Name']} ({row['Major']}): {row['Average']:.0f}"

df["Summary"] = df.apply(student_summary, axis=1)
print(df["Summary"])
````

### 11.3 `map()` for Value Mapping

Use `map()` to replace values based on a dictionary — perfect for recoding categories.

````python
# Map: replace values using a dictionary
major_full = {
    "DS": "Data Science",
    "Stats": "Statistics",
    "Math": "Mathematics",
    "CS": "Computer Science",
}

df["Major_Full"] = df["Major"].map(major_full)
print(df[["Name", "Major", "Major_Full"]])
````

## 12. Practical Example: Student Performance Analysis

Let's build a complete analysis pipeline from scratch — simulating a realistic scenario.

**Step 1:** Generate a realistic student dataset.

````python
rng = np.random.default_rng(2026)
n = 50

# Generate data
names = [f"Student_{i+1:02d}" for i in range(n)]
ages = rng.integers(19, 25, size=n)
majors = rng.choice(["Data Science", "Statistics", "Mathematics", "CS"], size=n)
genders = rng.choice(["M", "F"], size=n)
midterms = rng.normal(75, 12, size=n).clip(0, 100).round(1)
finals = rng.normal(72, 15, size=n).clip(0, 100).round(1)
homework = rng.normal(80, 8, size=n).clip(0, 100).round(1)

# Introduce some missing values (realistic!)
midterms[rng.choice(n, 3, replace=False)] = np.nan
finals[rng.choice(n, 2, replace=False)] = np.nan

df = pd.DataFrame({
    "Name": names,
    "Age": ages,
    "Gender": genders,
    "Major": majors,
    "Midterm": midterms,
    "Final": finals,
    "Homework": homework,
})

print(f"Dataset shape: {df.shape}")
df.head(10)
````

**Step 2:** Inspect the data — check for missing values and basic statistics.

````python
print("=== Data Info ===")
df.info()

print("\n=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Descriptive Statistics ===")
print(df.describe().round(2))
````

**Step 3:** Handle missing values — fill with the column median.

````python
# Fill missing exam scores with column median
print(f"Missing before: Midterm={df['Midterm'].isnull().sum()}, Final={df['Final'].isnull().sum()}")

df["Midterm"] = df["Midterm"].fillna(df["Midterm"].median())
df["Final"] = df["Final"].fillna(df["Final"].median())

print(f"Missing after:  Midterm={df['Midterm'].isnull().sum()}, Final={df['Final'].isnull().sum()}")
````

**Step 4:** Feature engineering — calculate weighted average and assign letter grades.

````python
# Weighted average: Midterm 30%, Final 50%, Homework 20%
df["Weighted_Avg"] = (
    df["Midterm"] * 0.30 +
    df["Final"] * 0.50 +
    df["Homework"] * 0.20
).round(2)

# Letter grade
def get_grade(score):
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

df["Grade"] = df["Weighted_Avg"].apply(get_grade)

print(df[["Name", "Midterm", "Final", "Homework", "Weighted_Avg", "Grade"]].head(10))
````

**Step 5:** Analyze grade distribution overall.

````python
# Grade distribution
grade_dist = df["Grade"].value_counts().sort_index()
print("=== Grade Distribution ===")
for grade, count in grade_dist.items():
    pct = count / len(df) * 100
    bar = "█" * count
    print(f"  {grade}: {count:>2} ({pct:>5.1f}%) {bar}")

print(f"\nClass average: {df['Weighted_Avg'].mean():.1f}")
print(f"Pass rate (≥60): {(df['Weighted_Avg'] >= 60).mean()*100:.1f}%")
````

**Step 6:** GroupBy analysis — compare performance across majors.

````python
# Performance by Major
major_stats = df.groupby("Major").agg(
    N=("Name", "count"),
    Avg_Score=("Weighted_Avg", "mean"),
    Std_Score=("Weighted_Avg", "std"),
    Min_Score=("Weighted_Avg", "min"),
    Max_Score=("Weighted_Avg", "max"),
).round(1)

print("=== Performance by Major ===")
print(major_stats)
````

**Step 7:** Grade distribution by major — cross-tabulation.

````python
# Cross-tabulation: Grade vs Major
cross_tab = pd.crosstab(df["Major"], df["Grade"])
print("=== Grade Distribution by Major ===")
print(cross_tab)
````

**Step 8:** Top performers and students at risk.

````python
# Top 5 students
print("=== Top 5 Students ===")
top5 = df.nlargest(5, "Weighted_Avg")[["Name", "Major", "Weighted_Avg", "Grade"]]
print(top5.to_string(index=False))

# Students at risk (grade D or F)
at_risk = df[df["Grade"].isin(["D", "F"])][["Name", "Major", "Weighted_Avg", "Grade"]]
print(f"\n=== Students at Risk ({len(at_risk)} students) ===")
print(at_risk.sort_values("Weighted_Avg").to_string(index=False))
````

**Step 9:** Correlation between exams — do Midterm scores predict Final scores?

````python
# Correlation matrix
corr = df[["Midterm", "Final", "Homework", "Weighted_Avg"]].corr().round(3)
print("=== Correlation Matrix ===")
print(corr)
````

**Step 10:** Save the final results to a CSV file.

````python
# Save final results
output = df[["Name", "Age", "Gender", "Major",
             "Midterm", "Final", "Homework",
             "Weighted_Avg", "Grade"]]
output.to_csv("student_grades.csv", index=False)
print(f"Saved {len(output)} records to 'student_grades.csv'")
````
