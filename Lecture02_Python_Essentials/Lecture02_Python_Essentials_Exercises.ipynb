````python
# Exercise 4.1: Grade Report
class_data = {
    "Alice": 92, "Bob": 78, "Charlie": 85, "Diana": 95,
    "Eve": 67, "Frank": 88, "Grace": 72, "Henry": 90,
}

def get_grade(score):
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

print(f"{'Name':<12} {'Score':>6} {'Grade':>6}")
print("=" * 26)
for name, score in class_data.items():
    print(f"{name:<12} {score:>6} {get_grade(score):>6}")

avg = sum(class_data.values()) / len(class_data)
print(f"\nAverage: {avg:.1f}")
above = [n for n, s in class_data.items() if s > avg]
print(f"Above average: {', '.join(above)}")
````

````python
# Exercise 5.1: Statistics Calculator
import math

def analyze_scores(scores):
    n = len(scores)
    mean = sum(scores) / n
    sorted_s = sorted(scores)
    if n % 2 == 0:
        median = (sorted_s[n//2 - 1] + sorted_s[n//2]) / 2
    else:
        median = sorted_s[n//2]
    std = math.sqrt(sum((x - mean) ** 2 for x in scores) / n)
    return {"count": n, "mean": round(mean, 1), "median": median,
            "std": round(std, 1), "min": min(scores), "max": max(scores)}

scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]
print(analyze_scores(scores))
````

````python
# Exercise 7.1: Student Class
class StudentRecord:
    def __init__(self, name):
        self.name = name
        self.scores = []
    
    def add_score(self, score):
        if not isinstance(score, (int, float)) or score < 0 or score > 100:
            print(f"Invalid score: {score}")
            return
        self.scores.append(score)
    
    def average(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0
    
    def best_score(self):
        return max(self.scores) if self.scores else 0
    
    def grade(self):
        avg = self.average()
        if avg >= 90: return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        elif avg >= 60: return "D"
        else: return "F"
    
    def __str__(self):
        return f"StudentRecord('{self.name}', scores={self.scores})"

s = StudentRecord("Alice")
s.add_score(92)
s.add_score(85)
s.add_score(78)
s.add_score(150)
print(s)
print(f"Average: {s.average():.1f}, Best: {s.best_score()}, Grade: {s.grade()}")
````

````python
# Exercise 7.2: Custom Exception
class DuplicateStudentError(Exception):
    def __init__(self, name):
        super().__init__(f"Student '{name}' already exists")

class StudentNotFoundError(Exception):
    def __init__(self, name):
        super().__init__(f"Student '{name}' not found")

class GradeBook:
    def __init__(self):
        self.records = {}
    
    def add_student(self, name, score):
        if name in self.records:
            raise DuplicateStudentError(name)
        self.records[name] = score
        print(f"Added {name}: {score}")
    
    def get_score(self, name):
        if name not in self.records:
            raise StudentNotFoundError(name)
        return self.records[name]
    
    def remove_student(self, name):
        if name not in self.records:
            raise StudentNotFoundError(name)
        del self.records[name]
        print(f"Removed {name}")

gb = GradeBook()
gb.add_student("Alice", 92)
gb.add_student("Bob", 78)
print(f"Alice's score: {gb.get_score('Alice')}")

try:
    gb.add_student("Alice", 85)
except DuplicateStudentError as e:
    print(f"Error: {e}")

try:
    gb.get_score("Charlie")
except StudentNotFoundError as e:
    print(f"Error: {e}")
````
