## 1. Variables & Data Types

A **variable** is a name that refers to a value stored in memory.  
Python is **dynamically typed** — you don't declare types; Python infers them.

### 1.1 The Four Basic Types

### 📌 About `f"..."` and `type()`

Throughout this notebook, you'll see code like:

```python
print(f"height = {height}, type = {type(height)}")
```

A few things to know:

- **f-string** (formatted string literal): Prefix a string with `f` and put expressions inside `{}` — Python evaluates them and inserts the result into the string.
- **`type(x)`**: Returns the data type of variable `x` (e.g., `<class 'int'>`, `<class 'float'>`).
- **`print()`**: Outputs the result to the screen.

```python
name = "Alice"
print(f"My name is {name}")        # → My name is Alice
print(f"1 + 2 = {1 + 2}")          # → 1 + 2 = 3
print(f"type = {type(name)}")       # → type = <class 'str'>
```

````python
# Integer (int) — whole numbers, no decimal point
age = 22
num_students = 35
negative = -10
print(f"age = {age}, type = {type(age)}")
print(f"num_students = {num_students}, type = {type(num_students)}")
````

**Float** — numbers with a decimal point. Note that division (`/`) always returns a float.

````python
# Float — numbers with decimal point
height = 1.75
pi = 3.14159
temperature = -2.5
print(f"height = {height}, type = {type(height)}")
print(f"pi = {pi}, type = {type(pi)}")

# Note: division always returns float
result = 10 / 3
print(f"10 / 3 = {result}, type = {type(result)}")
````

**String (`str`)** — text data, enclosed in single quotes, double quotes, or triple quotes for multi-line.

````python
# String (str) — text, enclosed in quotes
name = "Alice"
greeting = 'Hello, World!'
paragraph = """This is a
multi-line
string."""

print(f"name = '{name}', type = {type(name)}")
print(f"greeting = '{greeting}'")
print(f"paragraph = '{paragraph}'")
````

**Boolean (`bool`)** — only two values: `True` or `False`. Booleans are the result of comparisons.

````python
# Boolean (bool) — True or False (capital T and F!)
is_student = True
has_passed = False
print(f"is_student = {is_student}, type = {type(is_student)}")
print(f"has_passed = {has_passed}, type = {type(has_passed)}")

# Booleans come from comparisons
print(f"5 > 3  = {5 > 3}")
print(f"5 == 3 = {5 == 3}")
print(f"5 != 3 = {5 != 3}")
````

### 1.2 Type Conversion

You can convert between types using `int()`, `float()`, `str()`, `bool()`.

````python
# int → float
x = 5
print(f"int to float: {x} → {float(x)}")

# float → int (truncates, does NOT round)
y = 3.7
print(f"float to int: {y} → {int(y)}")

# str → int / float
age_str = "22"
print(f"str to int: '{age_str}' → {int(age_str)}")

price_str = "9.99"
print(f"str to float: '{price_str}' → {float(price_str)}")

# number → str
score = 95
print(f"int to str: {score} → '{str(score)}'")

# bool conversions
print(f"\nbool(1) = {bool(1)}")
print(f"bool(0) = {bool(0)}")
print(f"bool('hello') = {bool('hello')}")
print(f"bool('') = {bool('')}")
print(f"bool([]) = {bool([])}")
print(f"bool([1,2]) = {bool([1,2])}")
````

### 1.3 Arithmetic Operators

````python
a, b = 17, 5

print(f"a = {a}, b = {b}")
print(f"a + b  = {a + b}    (addition)")
print(f"a - b  = {a - b}   (subtraction)")
print(f"a * b  = {a * b}   (multiplication)")
print(f"a / b  = {a / b}   (division → always float)")
print(f"a // b = {a // b}     (floor division → int)")
print(f"a % b  = {a % b}     (modulo → remainder)")
print(f"a ** b = {a ** b} (power → 17^5)")
````

### 1.4 Assignment Shortcuts

````python
# Assign multiple variables in one line
name, age, gpa = "Alice", 22, 3.8
print(f"{name}, age {age}, GPA {gpa}")

# Swap two variables — no temp variable needed!
a, b = 10, 20
print(f"Before swap: a={a}, b={b}")
a, b = b, a
print(f"After swap:  a={a}, b={b}")
````

### 1.6 Variable Naming Rules

| Rule | Good | Bad |
|------|------|-----|
| Start with letter or `_` | `score`, `_temp` | `2nd_score` |
| Only letters, numbers, `_` | `student_name` | `student-name` |
| Case-sensitive | `Age` ≠ `age` | — |
| Don't use Python keywords | `my_list` | `list`, `print`, `for` |
| Use descriptive names | `avg_score` | `x`, `a1` |
| Convention: `snake_case` | `num_students` | `numStudents` |

## 3. Data Structures

Python has 4 built-in collection types:

| Type | Ordered | Mutable | Duplicates | Syntax |
|------|---------|---------|------------|--------|
| **List** | Yes | Yes | Yes | `[1, 2, 3]` |

### 3.1 Lists — The Workhorse of Python

````python
# Creating lists
scores = [78, 92, 85, 67, 95, 88, 72, 90]
names = ["Alice", "Bob", "Charlie", "Diana"]
mixed = [1, "hello", 3.14, True]  # Can mix types (but rarely should)
empty = []

print(f"scores: {scores}")
print(f"names:  {names}")
print(f"Length: {len(scores)}")
````

````python
# Modifying lists
fruits = ["apple", "banana", "cherry"]
print(f"Original: {fruits}")

# Change an element
fruits[1] = "blueberry"
print(f"After change: {fruits}")

# Add elements
fruits.append("date")           # Add to end
print(f"After append: {fruits}")

fruits.insert(1, "avocado")     # Insert at index 1
print(f"After insert: {fruits}")

# Remove elements
fruits.remove("cherry")         # Remove by value
print(f"After remove: {fruits}")

popped = fruits.pop()            # Remove and return last
print(f"Popped: {popped}, list: {fruits}")

popped = fruits.pop(0)           # Remove and return index 0
print(f"Popped: {popped}, list: {fruits}")
````

### 3.2 Tuples — Immutable Lists

````python
# Tuples look like lists but use parentheses
point = (3, 4)
rgb = (255, 128, 0)
single = (42,)  # Note the comma for single-element tuple

print(f"point = {point}, type = {type(point)}")
print(f"x = {point[0]}, y = {point[1]}")

# Tuple unpacking
x, y = point
print(f"Unpacked: x={x}, y={y}")

r, g, b = rgb
print(f"RGB: r={r}, g={g}, b={b}")
````

````python
# Tuples are IMMUTABLE — you cannot change them
# point[0] = 10  # This would cause TypeError!

# Why use tuples?
# 1. Signal that data should not change (coordinates, dates, config)
# 2. Can be used as dictionary keys (lists cannot)
# 3. Slightly faster than lists

# Functions often return tuples
def min_max(data):
    return min(data), max(data)  # returns a tuple

lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"min={lo}, max={hi}")
````

### 3.3 Dictionaries — Key-Value Pairs

````python
# Creating dictionaries
student = {
    "name": "Alice",
    "age": 22,
    "major": "Data Science",
    "gpa": 3.8,
    "courses": ["Statistics", "ML", "Python"]
}

# Access by key
print(f"Name: {student['name']}")
print(f"GPA:  {student['gpa']}")
print(f"Courses: {student['courses']}")

# Safe access with .get() — returns None if key doesn't exist
print(f"Email: {student.get('email')}")
print(f"Email: {student.get('email', 'not provided')}")
````

````python
# Practical example: counting word frequency
text = "the cat sat on the mat the cat likes the mat"
words = text.split()

freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print(f"Word frequency: {freq}")
print(f"Most common: '{max(freq, key=freq.get)}' ({max(freq.values())} times)")
````

## 4. Control Flow

### 4.1 `if / elif / else`

````python
# Basic if-else
temperature = 35

if temperature >= 30:
    print("It's hot!")
elif temperature >= 20:
    print("It's warm.")
elif temperature >= 10:
    print("It's cool.")
else:
    print("It's cold!")
````

````python
# Grade classifier — you'll use this pattern a lot
def get_grade(score):
    """Convert numeric score to letter grade."""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Test it
for score in [95, 87, 73, 62, 45]:
    print(f"Score {score} → Grade {get_grade(score)}")
````

````python
# Logical operators: and, or, not
age = 22
gpa = 3.5

# Must meet both conditions
if age >= 18 and gpa >= 3.0:
    print("Eligible for scholarship")

# Must meet at least one condition
if gpa >= 3.8 or age < 20:
    print("Special consideration")
else:
    print("Standard application")

# Negation
is_absent = False
if not is_absent:
    print("Student is present")
````

````python
# Ternary expression (one-line if-else)
score = 85
result = "Pass" if score >= 60 else "Fail"
print(f"Score {score}: {result}")

# Useful for quick assignments
x = -5
abs_x = x if x >= 0 else -x
print(f"Absolute value of {x} is {abs_x}")
````

````python
# Practical: calculate class statistics
scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]

total = 0
count_pass = 0
count_fail = 0

for score in scores:
    total += score
    if score >= 60:
        count_pass += 1
    else:
        count_fail += 1

mean = total / len(scores)
print(f"Total students: {len(scores)}")
print(f"Mean score:     {mean:.1f}")
print(f"Passed:         {count_pass}")
print(f"Failed:         {count_fail}")
print(f"Pass rate:      {count_pass/len(scores)*100:.0f}%")
````

### 4.3 `while` Loop

````python
# while: keep looping until condition is False
# Example: how many times can you halve a number before it's < 1?
n = 1000
count = 0

while n >= 1:
    n /= 2
    count += 1

print(f"Halved {count} times to get below 1")
print(f"Final value: {n:.6f}")
````

````python
# Newton's method: find square root of 2
target = 2
guess = 1.0
tolerance = 1e-10
iteration = 0

while abs(guess * guess - target) > tolerance:
    guess = (guess + target / guess) / 2
    iteration += 1
    print(f"Iteration {iteration}: guess = {guess:.15f}")

print(f"\nSquare root of {target} ≈ {guess}")
print(f"Converged in {iteration} iterations")
````

### 4.4 `break` and `continue`

````python
# break: exit the loop immediately
# Find the first score below 70
scores = [88, 92, 85, 65, 78, 55, 90]

for i, score in enumerate(scores):
    if score < 70:
        print(f"First failing score: {score} at index {i}")
        break
else:
    # This runs only if the loop completes without break
    print("All students passed!")
````

````python
# continue: skip the rest of this iteration
# Print only passing scores
scores = [88, 92, 55, 85, 42, 78, 90]

print("Passing scores:")
for score in scores:
    if score < 60:
        continue  # Skip this score
    print(f"  {score}")
````

### 5.2 Default Arguments

````python
def calculate_mean(data, trim=0):
    """
    Calculate mean with optional trimming.
    trim: number of values to remove from each end.
    """
    sorted_data = sorted(data)
    if trim > 0:
        sorted_data = sorted_data[trim:-trim]
    return sum(sorted_data) / len(sorted_data)

scores = [10, 78, 82, 85, 88, 90, 92, 95, 100]
print(f"Regular mean:  {calculate_mean(scores):.1f}")
print(f"Trimmed mean:  {calculate_mean(scores, trim=1):.1f}")  # Remove lowest & highest
print(f"Trimmed mean:  {calculate_mean(scores, trim=2):.1f}")  # Remove 2 from each end
````

### 5.3 Returning Multiple Values

````python
def describe(data):
    """Return basic statistics of a list."""
    n = len(data)
    mean = sum(data) / n
    sorted_data = sorted(data)
    
    # Median
    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]
    
    # Standard deviation
    variance = sum((x - mean) ** 2 for x in data) / n
    std = variance ** 0.5
    
    return mean, median, std, min(data), max(data)

# Unpack the returned tuple
scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]
mean, median, std, lo, hi = describe(scores)

print(f"Scores: {scores}")
print(f"Mean:   {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Std:    {std:.2f}")
print(f"Min:    {lo}")
print(f"Max:    {hi}")
````

### 5.4 `*args` and `**kwargs`

````python
# *args: accept any number of positional arguments
def add_all(*numbers):
    """Sum any number of arguments."""
    return sum(numbers)

print(f"add_all(1, 2):       {add_all(1, 2)}")
print(f"add_all(1, 2, 3, 4): {add_all(1, 2, 3, 4)}")
````

````python
# **kwargs: accept any number of keyword arguments
def create_student(**info):
    """Create a student record from keyword arguments."""
    for key, value in info.items():
        print(f"  {key}: {value}")

print("Student 1:")
create_student(name="Alice", age=22, gpa=3.8)

print("Student 2:")
create_student(name="Bob", age=21, major="Statistics", year=3)

# You can also pass a dictionary using ** to unpack it
print("\nStudent 3 (from dict):")
student_data = {"name": "Charlie", "age": 23, "gpa": 3.6, "major": "Data Science"}
create_student(**student_data)
````

### 5.5 Lambda Functions (Anonymous Functions)

````python
# Lambda: a small one-line function
# Syntax: lambda arguments: expression

square = lambda x: x ** 2
print(f"square(5) = {square(5)}")

add = lambda a, b: a + b
print(f"add(3, 4) = {add(3, 4)}")
````

````python
# Lambda is most useful with sorted(), map(), filter()

# Sort students by GPA (descending)
students = [
    {"name": "Alice", "gpa": 3.8},
    {"name": "Bob", "gpa": 3.2},
    {"name": "Charlie", "gpa": 3.9},
    {"name": "Diana", "gpa": 3.5},
]

sorted_by_gpa = sorted(students, key=lambda s: s["gpa"], reverse=True)
for s in sorted_by_gpa:
    print(f"{s['name']:10s} GPA: {s['gpa']}")
````

````python
# map(): apply a function to every element
scores = [78, 92, 85, 67, 95]
grades = list(map(get_grade, scores))
print(f"Scores: {scores}")
print(f"Grades: {grades}")

# filter(): keep only elements that satisfy a condition
high_scores = list(filter(lambda x: x >= 85, scores))
print(f"High scores (≥85): {high_scores}")
````

## 6. List Comprehension

List comprehension is a **concise, Pythonic way** to create lists. Master this — it's everywhere in data science code.

### 6.1 Basic Syntax

```python
new_list = [expression for item in iterable]
```

````python
# Squares of 1 to 10
squares = [x ** 2 for x in range(1, 11)]
print(f"Squares: {squares}")

# Convert temperatures: Celsius to Fahrenheit
celsius = [0, 10, 20, 30, 40]
fahrenheit = [c * 9/5 + 32 for c in celsius]
print(f"Celsius:    {celsius}")
print(f"Fahrenheit: {fahrenheit}")

# Uppercase all names
names = ["alice", "bob", "charlie"]
upper_names = [name.upper() for name in names]
print(f"Upper: {upper_names}")
````

### 6.2 With Condition

```python
new_list = [expression for item in iterable if condition]
```

````python
scores = [78, 92, 85, 67, 95, 88, 72, 90, 45, 83]

# Only passing scores
passing = [s for s in scores if s >= 60]
print(f"Passing: {passing}")

# Only even numbers from 1 to 20
evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"Evens: {evens}")

# Words longer than 4 characters
words = ["Python", "is", "great", "for", "statistics"]
long_words = [w for w in words if len(w) > 4]
print(f"Long words: {long_words}")
````

### 6.3 With if-else

```python
new_list = [expr_if_true if condition else expr_if_false for item in iterable]
```

````python
scores = [78, 92, 85, 67, 95, 88, 72, 90, 45, 83]

# Label each score as Pass or Fail
labels = ["Pass" if s >= 60 else "Fail" for s in scores]
print(f"Scores: {scores}")
print(f"Labels: {labels}")

# Grade each score
grades = [get_grade(s) for s in scores]
print(f"Grades: {grades}")
````

### 6.4 Dictionary & Set Comprehension

````python
# Dictionary comprehension
names = ["Alice", "Bob", "Charlie"]
scores = [92, 78, 85]

grade_book = {name: score for name, score in zip(names, scores)}
print(f"Grade book: {grade_book}")

# Square lookup table
squares = {x: x**2 for x in range(1, 11)}
print(f"Squares: {squares}")

# Set comprehension
sentence = "the quick brown fox jumps over the lazy dog"
unique_lengths = {len(word) for word in sentence.split()}
print(f"Unique word lengths: {unique_lengths}")
````

### 6.5 Nested Comprehension

````python
# Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print(f"Matrix: {matrix}")
print(f"Flat:   {flat}")

# Create a multiplication table
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
print("\nMultiplication table:")
for row in table:
    print(row)
````

## 7. File I/O — Reading and Writing Files

Data comes from files. Let's learn how to read and write them.

### 7.1 Writing a Text File

````python
# Write a CSV file
data = [
    ["Name", "Age", "Score"],
    ["Alice", 22, 92],
    ["Bob", 21, 78],
    ["Charlie", 23, 85],
    ["Diana", 22, 95],
    ["Eve", 20, 67],
]

with open("students.csv", "w") as f:
    for row in data:
        line = ",".join(str(item) for item in row)
        f.write(line + "\n")

print("File 'students.csv' written successfully!")
````

### 7.2 Reading a Text File

````python
# Method 1: Read entire file
with open("students.csv", "r") as f:
    content = f.read()
print("--- Full content ---")
print(content)
````

````python
# Method 2: Read line by line and parse
students = []

with open("students.csv", "r") as f:
    header = f.readline().strip().split(",")
    for line in f:
        name, age, score = line.strip().split(",")
        students.append({
            "name": name,
            "age": int(age),
            "score": int(score)
        })

print(f"Header: {header}")
print(f"Loaded {len(students)} students:")
for s in students:
    print(f"  {s['name']:10s} age {s['age']}, score {s['score']}")

# Calculate average
avg = sum(s["score"] for s in students) / len(students)
print(f"\nAverage score: {avg:.1f}")
````

### 7.3 The `with` Statement


## 8. Classes & Encapsulation (OOP Basics)

A **class** is a blueprint for creating objects. It bundles **data** (attributes) and **behavior** (methods) together.  
This is called **encapsulation** — one of the core ideas of Object-Oriented Programming (OOP).

### 8.1 Defining a Class

````python
# A simple class
class Student:
    """A class representing a student."""
    
    def __init__(self, name, age, major):
        """Initialize student attributes."""
        self.name = name      # instance attribute
        self.age = age
        self.major = major
        self.courses = []     # default empty list
    
    def enroll(self, course):
        """Add a course to the student's course list."""
        self.courses.append(course)
        print(f"{self.name} enrolled in {course}")
    
    def info(self):
        """Display student information."""
        print(f"Name: {self.name}, Age: {self.age}, Major: {self.major}")
        print(f"Courses: {self.courses}")

# Create objects (instances)
alice = Student("Alice", 22, "Data Science")
bob = Student("Bob", 21, "Statistics")

alice.enroll("Applied Statistics")
alice.enroll("Machine Learning")
alice.info()

print()
bob.enroll("Linear Algebra")
bob.info()
````

### 8.2 Encapsulation — Bundling Data and Methods

Encapsulation means keeping data and the methods that operate on it **inside one unit** (the class).  
This prevents outside code from accidentally breaking internal state.

````python
class BankAccount:
    """A bank account with controlled access to balance."""
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance  # convention: _ means "internal, don't touch directly"
    
    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive!")
            return
        self._balance += amount
        print(f"Deposited ${amount}. Balance: ${self._balance}")
    
    def withdraw(self, amount):
        if amount > self._balance:
            print(f"Insufficient funds! Balance: ${self._balance}")
            return
        self._balance -= amount
        print(f"Withdrew ${amount}. Balance: ${self._balance}")
    
    def get_balance(self):
        return self._balance

account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # insufficient funds
print(f"Final balance: ${account.get_balance()}")
````

### 8.3 Special Methods (`__str__`, `__repr__`)

Special methods (also called **dunder methods**) let your class work with Python built-in functions like `print()`, `len()`, etc.

````python
class Score:
    """A score with a nice string representation."""
    
    def __init__(self, student, value):
        self.student = student
        self.value = value
    
    def __str__(self):
        """Used by print()"""
        return f"{self.student}: {self.value} points"
    
    def __repr__(self):
        """Used in lists and debugging"""
        return f"Score('{self.student}', {self.value})"
    
    def grade(self):
        if self.value >= 90: return "A"
        elif self.value >= 80: return "B"
        elif self.value >= 70: return "C"
        elif self.value >= 60: return "D"
        else: return "F"

s = Score("Alice", 92)
print(s)         # calls __str__
print(repr(s))   # calls __repr__
print(f"Grade: {s.grade()}")

# In a list, __repr__ is used
scores = [Score("Alice", 92), Score("Bob", 78), Score("Charlie", 85)]
print(scores)
````

### 8.4 Inheritance — Building on Existing Classes

A child class can **inherit** all attributes and methods from a parent class, and add or override its own.

````python
class Person:
    """Base class for all people."""
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"{self.name} (age {self.age})"


class GradStudent(Person):
    """A graduate student — inherits from Person."""
    def __init__(self, name, age, research_topic):
        super().__init__(name, age)  # call parent's __init__
        self.research_topic = research_topic
    
    def __str__(self):
        return f"{self.name} (age {self.age}), researching: {self.research_topic}"


p = Person("Dr. Smith", 45)
g = GradStudent("Alice", 25, "Bayesian Statistics")

print(p)
print(g)
print(f"Is GradStudent a Person? {isinstance(g, Person)}")
````

## 9. Error Handling

Errors happen. Good code handles them gracefully.

### 9.1 Common Errors

````python
# Let's see some common errors (uncomment one at a time to see)

# TypeError: wrong type
# "hello" + 5

# ValueError: right type but wrong value
# int("hello")

# IndexError: list index out of range
# [1, 2, 3][5]

# KeyError: dictionary key not found
# {"name": "Alice"}["age"]

# ZeroDivisionError
# 10 / 0

# FileNotFoundError
# open("nonexistent.txt")

print("Uncomment any line above to see the error.")
````

### 9.2 `try / except`

````python
# Basic try-except
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None

print(f"10 / 3 = {safe_divide(10, 3):.2f}")
print(f"10 / 0 = {safe_divide(10, 0)}")
````

````python
# Practical: safely convert user input
def parse_score(text):
    """Convert text to score, handling errors."""
    try:
        score = float(text)
        if 0 <= score <= 100:
            return score
        else:
            print(f"Warning: Score {score} is out of range [0, 100]")
            return None
    except ValueError:
        print(f"Error: '{text}' is not a valid number")
        return None

# Test
test_inputs = ["85", "92.5", "hello", "150", "-10"]
for t in test_inputs:
    result = parse_score(t)
    print(f"  '{t}' → {result}")
    print()
````

````python
# Multiple except blocks + finally
def read_score_from_file(filename):
    try:
        with open(filename, "r") as f:
            text = f.read().strip()
            score = float(text)
            return score
    except FileNotFoundError:
        print(f"File '{filename}' not found")
    except ValueError:
        print(f"File content is not a valid number")
    except Exception as e:
        print(f"Unexpected error: {e}")
    finally:
        print("(Read attempt completed)")
    return None

result = read_score_from_file("nonexistent.txt")
````

### 9.3 Custom Exceptions

You can define your own exception classes by inheriting from `Exception`.  
This is where **classes** meet **error handling** — custom exceptions make your code more readable and specific.

````python
# Define custom exceptions
class ScoreError(Exception):
    """Base exception for score-related errors."""
    pass

class ScoreOutOfRangeError(ScoreError):
    """Raised when score is not between 0 and 100."""
    def __init__(self, score):
        self.score = score
        super().__init__(f"Score {score} is out of range (0-100)")

class InvalidScoreTypeError(ScoreError):
    """Raised when score is not a number."""
    def __init__(self, value):
        self.value = value
        super().__init__(f"Expected a number, got {type(value).__name__}: {value}")


# Use custom exceptions
def validate_score(score):
    """Validate a score value, raising custom exceptions."""
    if not isinstance(score, (int, float)):
        raise InvalidScoreTypeError(score)
    if score < 0 or score > 100:
        raise ScoreOutOfRangeError(score)
    return score


# Test with different inputs
test_values = [85, 150, -10, "hello", 92]

for val in test_values:
    try:
        result = validate_score(val)
        print(f"✓ {val} is valid")
    except ScoreError as e:
        print(f"✗ {e}")
````

````python
# Combining classes with custom exceptions — a practical example
class GradeBook:
    """A grade book that validates scores before adding them."""
    
    def __init__(self):
        self.records = {}
    
    def add_score(self, name, score):
        try:
            validate_score(score)
            self.records[name] = score
            print(f"Added: {name} = {score}")
        except ScoreError as e:
            print(f"Cannot add score for {name}: {e}")
    
    def average(self):
        if not self.records:
            return 0
        return sum(self.records.values()) / len(self.records)
    
    def __str__(self):
        return f"GradeBook({len(self.records)} students, avg={self.average():.1f})"


gb = GradeBook()
gb.add_score("Alice", 92)
gb.add_score("Bob", 78)
gb.add_score("Charlie", 150)    # out of range
gb.add_score("Diana", "good")   # wrong type
gb.add_score("Eve", 85)

print(f"\n{gb}")
````
