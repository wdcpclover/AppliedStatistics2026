## 9. Variables & Data Types

A **variable** stores a value in memory. Python is **dynamically typed** — you don't need to declare the type; Python figures it out.

### 9.1 Basic Data Types

````python
# Integer (int) — whole numbers
num_students = 35
print(f"num_students = {num_students}, type = {type(num_students)}")

# Float — decimal numbers
avg_score = 85.6
print(f"avg_score = {avg_score}, type = {type(avg_score)}")

# String (str) — text
course_name = "Applied Statistics"
print(f"course_name = '{course_name}', type = {type(course_name)}")

# Boolean (bool) — True or False
is_fun = True
print(f"is_fun = {is_fun}, type = {type(is_fun)}")
````

### 9.2 Collections: Lists

A **list** is an ordered collection of items — you'll use lists everywhere in data science.

````python
# A list of exam scores
scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]
print(f"Scores: {scores}")
print(f"Number of students: {len(scores)}")
print(f"First score: {scores[0]}")
print(f"Last score: {scores[-1]}")
print(f"Top 3 scores: {scores[:3]}")
````

````python
# Simple statistics — without any library!
total = sum(scores)
count = len(scores)
mean = total / count
maximum = max(scores)
minimum = min(scores)

print(f"Sum:  {total}")
print(f"Mean: {mean}")
print(f"Max:  {maximum}")
print(f"Min:  {minimum}")

# Sorted
print(f"Sorted: {sorted(scores)}")
````

### 9.3 Dictionaries: Key-Value Pairs

A **dictionary** maps keys to values — like a real dictionary maps words to definitions.

````python
# Student record as a dictionary
student = {
    "name": "Alice",
    "age": 22,
    "major": "Data Science",
    "gpa": 3.8
}

print(f"Name:  {student['name']}")
print(f"Major: {student['major']}")
print(f"GPA:   {student['gpa']}")
````

## 10. Control Flow: Making Decisions & Repeating

### 10.1 `if / elif / else` — Conditional Logic

````python
# Grade classification based on score
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

### 10.2 `for` Loop — Repeating Actions

````python
# Calculate the grade for each student
scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]

for i, score in enumerate(scores):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    print(f"Student {i+1}: {score} → {grade}")
````

### 10.3 List Comprehension — Pythonic Short Loops

List comprehension is a concise way to create lists. It's very common in Python.

````python
# Traditional loop
squares = []
for x in range(1, 11):
    squares.append(x ** 2)
print("Loop:         ", squares)

# List comprehension — same result, one line!
squares = [x ** 2 for x in range(1, 11)]
print("Comprehension:", squares)

# With a condition: only even squares
even_squares = [x ** 2 for x in range(1, 11) if x % 2 == 0]
print("Even squares: ", even_squares)
````

## 11. Functions: Reusable Code Blocks

A **function** groups code into a reusable block. You call it by name whenever you need it.

````python
def calculate_mean(data):
    """Calculate the arithmetic mean of a list of numbers."""
    return sum(data) / len(data)


def calculate_std(data):
    """Calculate the standard deviation of a list of numbers."""
    mean = calculate_mean(data)
    squared_diffs = [(x - mean) ** 2 for x in data]
    variance = sum(squared_diffs) / len(data)
    return variance ** 0.5


# Test with our scores
scores = [78, 92, 85, 67, 95, 88, 72, 90, 83, 76]

print(f"Scores: {scores}")
print(f"Mean:   {calculate_mean(scores):.2f}")
print(f"Std:    {calculate_std(scores):.2f}")
````

> **Think about it**: We just computed the mean and standard deviation from scratch!  
> In the coming weeks, you'll learn how libraries like NumPy and Pandas do this in one line — but understanding the logic behind it is what makes you a *statistician*, not just a *coder*.

## 12. Your First Data Visualization

Let's bring data to life. We'll use **Matplotlib** — Python's most widely used plotting library.

### 12.1 A Simple Line Plot

````python
import matplotlib.pyplot as plt

# Monthly average temperatures in a city (°C)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temps = [2, 5, 10, 16, 22, 28, 31, 30, 25, 18, 10, 4]

plt.figure(figsize=(10, 5))
plt.plot(months, temps, marker="o", color="#e74c3c", linewidth=2)
plt.title("Monthly Average Temperature", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

### 12.2 A Bar Chart

````python
# Programming language popularity (fictional survey)
languages = ["Python", "R", "SQL", "Java", "Julia"]
popularity = [85, 45, 60, 35, 15]
colors = ["#3498db", "#2ecc71", "#e74c3c", "#f39c12", "#9b59b6"]

plt.figure(figsize=(8, 5))
bars = plt.bar(languages, popularity, color=colors, edgecolor="white", linewidth=1.5)
plt.title("Programming Language Popularity in Data Science", fontsize=14)
plt.ylabel("Popularity Score")

# Add value labels on top of bars
for bar, val in zip(bars, popularity):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1,
             str(val), ha="center", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
````

### 12.3 A Histogram — The Statistician's Best Friend

A **histogram** shows the distribution of data. This is the most important chart in statistics.

````python
import numpy as np

# Generate 1000 random exam scores (normal distribution)
np.random.seed(42)  # for reproducibility
exam_scores = np.random.normal(loc=75, scale=10, size=1000)

plt.figure(figsize=(10, 5))
plt.hist(exam_scores, bins=30, color="#3498db", edgecolor="white", alpha=0.8)
plt.axvline(x=np.mean(exam_scores), color="red", linestyle="--", linewidth=2, label=f"Mean = {np.mean(exam_scores):.1f}")
plt.title("Distribution of Exam Scores (n=1000)", fontsize=16)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Mean:   {np.mean(exam_scores):.2f}")
print(f"Std:    {np.std(exam_scores):.2f}")
print(f"Min:    {np.min(exam_scores):.2f}")
print(f"Max:    {np.max(exam_scores):.2f}")
````

> **This is a Normal Distribution** — the most important distribution in statistics.  
> We generated 1000 scores with mean=75 and std=10. Notice the bell shape?  
> We will study this in depth in **Week 6: Probability & Distributions**.

### 12.4 A Scatter Plot — Exploring Relationships

````python
# Study hours vs exam score
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 50)
exam_score = 50 + 4 * study_hours + np.random.normal(0, 5, 50)

plt.figure(figsize=(8, 6))
plt.scatter(study_hours, exam_score, color="#3498db", alpha=0.7, edgecolors="white", s=80)

# Add a trend line
z = np.polyfit(study_hours, exam_score, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 10, 100)
plt.plot(x_line, p(x_line), color="#e74c3c", linewidth=2, linestyle="--", label=f"Trend: y = {z[0]:.1f}x + {z[1]:.1f}")

plt.title("Study Hours vs Exam Score", fontsize=16)
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

> **Question for the class:** What does this chart tell us?  
> - Is there a relationship between study hours and exam scores?  
> - Is the relationship positive or negative?  
> - Can we predict a score given the study hours?  
>
> We will answer these questions formally in **Week 9: Correlation & Regression**.

## 13. Putting It All Together: A Mini Data Analysis

Let's combine everything we've learned into a small end-to-end analysis.

**Scenario:** You surveyed 20 students about their daily coffee consumption and exam scores.

````python
import pandas as pd

# Create a dataset
np.random.seed(2026)
n = 20

data = pd.DataFrame({
    "Student": [f"S{i+1:02d}" for i in range(n)],
    "Coffee_Cups": np.random.randint(0, 6, n),
    "Study_Hours": np.round(np.random.uniform(1, 8, n), 1),
    "Exam_Score": np.random.randint(55, 100, n)
})

# Display the first 10 rows
data.head(10)
````

````python
# Quick summary statistics — one line with Pandas!
data.describe()
````

````python
# Visualize: Create a dashboard with 3 charts
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Chart 1: Score distribution
axes[0].hist(data["Exam_Score"], bins=8, color="#3498db", edgecolor="white")
axes[0].set_title("Score Distribution")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Count")

# Chart 2: Study hours vs Score
axes[1].scatter(data["Study_Hours"], data["Exam_Score"], color="#2ecc71", s=80, edgecolors="white")
axes[1].set_title("Study Hours vs Score")
axes[1].set_xlabel("Study Hours")
axes[1].set_ylabel("Exam Score")

# Chart 3: Coffee consumption vs Score
axes[2].scatter(data["Coffee_Cups"], data["Exam_Score"], color="#e74c3c", s=80, edgecolors="white")
axes[2].set_title("Coffee Cups vs Score")
axes[2].set_xlabel("Coffee Cups / Day")
axes[2].set_ylabel("Exam Score")

plt.suptitle("Student Performance Analysis", fontsize=16, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()
````

````python
# Correlation: Are these variables related?
print("Correlation Matrix:")
print(data[["Coffee_Cups", "Study_Hours", "Exam_Score"]].corr().round(3))
````

> **Discussion:**
> - What patterns do you see in the charts?
> - Does coffee consumption correlate with exam scores? Does it mean coffee *causes* better scores?
> - What's the difference between **correlation** and **causation**?
>
> These are exactly the kinds of questions this course will teach you to answer rigorously.
